```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
B12 Deficiency Prediction Model - Stage 1 Logistic Regression Model
Predicting B12 status based on clinical features and DNA methylation features

Author: Research Team
Date: August 23, 2025
Version: 1.0 - Fixed Data Leakage Issues - Logistic Regression Implementation

Feature Set: Clinical features (excluding B12_mol) + DNA methylation promoter region CpG sites
Objective: Predict B12 deficiency status (LB vs NB) without data leakage using Logistic Regression
"""

import os
import sys
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)

# Machine learning libraries
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import (train_test_split, StratifiedKFold, 
                                   cross_val_score, GridSearchCV)
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import (classification_report, confusion_matrix, 
                           roc_auc_score, roc_curve, precision_recall_curve,
                           f1_score, accuracy_score, matthews_corrcoef,
                           cohen_kappa_score, brier_score_loss)
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json
from datetime import datetime
import logging

class B12PredictionLR:
    """B12 deficiency prediction Logistic Regression model class - Fixed Data Leakage"""
    
    def __init__(self, output_dir="/Users/heweilin/Desktop/P056_Code_3/Data", prefix="LR_"):
        self.output_dir = output_dir
        self.prefix = prefix
        self.timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        
        os.makedirs(output_dir, exist_ok=True)
        self.setup_logging()
        
        # Initialize data storage
        self.clinical_data = None
        self.methylation_data = None
        self.dmr_data = None
        self.X = None
        self.y = None
        self.X_train = None
        self.X_test = None
        self.X_val = None
        self.y_train = None
        self.y_test = None
        self.y_val = None
        self.feature_names = None
        self.model = None
        self.scaler = None
        self.label_encoder = None
        self.raw_clinical_features = None
        self.raw_methylation_data = None
        
        print("="*80)
        print("B12 DEFICIENCY PREDICTION MODEL - STAGE 1 LOGISTIC REGRESSION")
        print("FIXED DATA LEAKAGE VERSION")
        print("="*80)
        print(f"Timestamp: {self.timestamp}")
        print(f"Output directory: {output_dir}")
        print(f"File prefix: {prefix}")
        print("\nKey Features:")
        print("- L2 regularization with balanced class weights")
        print("- Liblinear solver optimized for small samples") 
        print("- Feature scaling with StandardScaler")
        print("- Comprehensive hyperparameter optimization")
        print("- Reproducible results with fixed random seeds")
        print("="*80)
    
    def setup_logging(self):
        log_file = os.path.join(self.output_dir, f"{self.prefix}model_log_{self.timestamp}.txt")
        logging.basicConfig(
            level=logging.INFO,
            format='%(asctime)s - %(levelname)s - %(message)s',
            handlers=[
                logging.FileHandler(log_file),
                logging.StreamHandler(sys.stdout)
            ]
        )
        self.logger = logging.getLogger(__name__)
        self.logger.info("Logging setup completed")
    
    def load_data(self, clinical_path, dmr_path, methylation_path):
        print("\n" + "="*50)
        print("STEP 1: DATA LOADING AND PREPROCESSING")
        print("="*50)
        
        self.logger.info("Starting data loading process")
        
        print(f"\n1.1 Loading clinical data from: {clinical_path}")
        try:
            self.clinical_data = pd.read_csv(clinical_path)
            print(f"    Clinical data shape: {self.clinical_data.shape}")
            print(f"    Columns: {list(self.clinical_data.columns)}")
            self.logger.info(f"Clinical data loaded successfully: {self.clinical_data.shape}")
        except Exception as e:
            print(f"    Error loading clinical data: {e}")
            self.logger.error(f"Error loading clinical data: {e}")
            return False
        
        print(f"\n1.2 Loading DMR data from: {dmr_path}")
        try:
            self.dmr_data = pd.read_csv(dmr_path, low_memory=False)
            print(f"    DMR data shape: {self.dmr_data.shape}")
            print(f"    Columns: {list(self.dmr_data.columns[:10])}...")
            self.logger.info(f"DMR data loaded successfully: {self.dmr_data.shape}")
        except Exception as e:
            print(f"    Error loading DMR data: {e}")
            self.logger.error(f"Error loading DMR data: {e}")
            return False
        
        print(f"\n1.3 Loading methylation data from: {methylation_path}")
        try:
            self.methylation_data = pd.read_csv(methylation_path)
            print(f"    Methylation data shape: {self.methylation_data.shape}")
            print(f"    Gene symbols: {self.methylation_data.shape[0]} genes")
            print(f"    Samples: {self.methylation_data.shape[1]-1} samples")
            self.logger.info(f"Methylation data loaded successfully: {self.methylation_data.shape}")
        except Exception as e:
            print(f"    Error loading methylation data: {e}")
            self.logger.error(f"Error loading methylation data: {e}")
            return False
        
        return True
    
    def preprocess_clinical_data(self):
        print("\n1.4 Preprocessing clinical data (EXCLUDING B12_mol to avoid target leakage)")
        
        print("    B12 status distribution:")
        b12_counts = self.clinical_data['B12_status'].value_counts()
        print(f"    {b12_counts}")
        
        clinical_features = ['age', 'BMI', 'parity', 'B12supplem', 
                           'v1p_MultivitTab', 'v1p_FolicAcid', 'smoking']
        
        print("    IMPORTANT: B12_mol EXCLUDED to prevent target leakage")
        print("    IMPORTANT: ethnicity EXCLUDED to avoid categorical encoding issues")
        
        available_features = []
        for feature in clinical_features:
            if feature in self.clinical_data.columns:
                available_features.append(feature)
            else:
                print(f"    Warning: {feature} not found in clinical data")
        
        print(f"    Available clinical features: {available_features}")
        
        clinical_X = self.clinical_data[available_features].copy()
        
        print("    Handling missing values in clinical data:")
        missing_counts = clinical_X.isnull().sum()
        print(f"    Missing values per feature:")
        for feature, count in missing_counts.items():
            if count > 0:
                print(f"      {feature}: {count} missing values")
        
        for col in clinical_X.columns:
            if clinical_X[col].isnull().sum() > 0:
                median_val = clinical_X[col].median()
                clinical_X[col].fillna(median_val, inplace=True)
                print(f"      Filled {col} with median: {median_val}")
        
        print("    Converting all features to numerical format:")
        for col in clinical_X.columns:
            clinical_X[col] = pd.to_numeric(clinical_X[col], errors='coerce')
            print(f"      {col}: {clinical_X[col].dtype}")
        
        if clinical_X.isnull().sum().sum() > 0:
            print("    Handling any NaN values from conversion:")
            clinical_X = clinical_X.fillna(clinical_X.median())
        
        print(f"    Final clinical data shape: {clinical_X.shape}")
        
        return clinical_X, available_features
    
    def prepare_methylation_data(self):
        print("\n1.5 Preparing methylation data")
        
        methylation_df = self.methylation_data.set_index('Gene_Symbol').T
        print(f"    Methylation data after transpose: {methylation_df.shape}")
        
        missing_counts = methylation_df.isnull().sum().sum()
        if missing_counts > 0:
            print(f"    Found {missing_counts} missing values in methylation data")
            methylation_df = methylation_df.fillna(methylation_df.median())
            print("    Missing values filled with median")
        
        clinical_ids = self.clinical_data['DNA_ID'].values
        methylation_samples = methylation_df.index.values
        
        print(f"    Clinical DNA IDs: {len(clinical_ids)}")
        print(f"    Methylation samples: {len(methylation_samples)}")
        
        common_samples = list(set(clinical_ids) & set(methylation_samples))
        print(f"    Common samples: {len(common_samples)}")
        
        if len(common_samples) == 0:
            print("    Warning: No common samples found between clinical and methylation data")
            return None
        
        methylation_filtered = methylation_df.loc[common_samples]
        print(f"    Filtered methylation data: {methylation_filtered.shape}")
        
        return methylation_filtered, common_samples
    
    def prepare_initial_dataset(self):
        print("\n1.6 Combining initial features")
        
        clinical_X, clinical_features = self.preprocess_clinical_data()
        
        methylation_result = self.prepare_methylation_data()
        if methylation_result is None:
            print("    Using only clinical features")
            self.X = clinical_X
            self.feature_names = clinical_features
            self.y = self.clinical_data['B12_status'].values
            self.raw_methylation_data = None
        else:
            methylation_X, common_samples = methylation_result
            print("    Combining clinical and methylation features")
            
            clinical_filtered = self.clinical_data[self.clinical_data['DNA_ID'].isin(common_samples)]
            clinical_filtered = clinical_filtered.set_index('DNA_ID').loc[common_samples]
            clinical_X_filtered = clinical_filtered[clinical_features]
            
            methylation_X_aligned = methylation_X.loc[common_samples]
            
            self.X = pd.concat([clinical_X_filtered, methylation_X_aligned], axis=1)
            self.feature_names = clinical_features + list(methylation_X_aligned.columns)
            
            self.raw_methylation_data = methylation_X_aligned
            self.raw_clinical_features = clinical_X_filtered
            
            self.y = clinical_filtered['B12_status'].values
        
        print(f"    Initial combined data shape: {self.X.shape}")
        print(f"    Total initial features: {len(self.feature_names)}")
        
        self.label_encoder = LabelEncoder()
        self.y = self.label_encoder.fit_transform(self.y)
        print(f"    Target encoding: {dict(zip(self.label_encoder.classes_, self.label_encoder.transform(self.label_encoder.classes_)))}")
        
        return True
    
    def split_data(self, test_size=0.25, val_size=0.15, random_state=42):
        print("\n" + "="*50)
        print("STEP 2: DATA SPLITTING (25% TEST SET)")
        print("="*50)
        
        print(f"Total samples: {len(self.X)}")
        print(f"Test size: {test_size} ({int(len(self.X) * test_size)} samples)")
        print(f"Validation size: {val_size} ({int(len(self.X) * val_size)} samples)")
        print(f"Training size: {1 - test_size - val_size} ({int(len(self.X) * (1 - test_size - val_size))} samples)")
        
        X_temp, self.X_test, y_temp, self.y_test = train_test_split(
            self.X, self.y, test_size=test_size, random_state=random_state, 
            stratify=self.y
        )
        
        val_size_adjusted = val_size / (1 - test_size)
        self.X_train, self.X_val, self.y_train, self.y_val = train_test_split(
            X_temp, y_temp, test_size=val_size_adjusted, random_state=random_state, 
            stratify=y_temp
        )
        
        print(f"\nFinal data splits:")
        print(f"    Training set: {self.X_train.shape[0]} samples")
        print(f"    Validation set: {self.X_val.shape[0]} samples") 
        print(f"    Test set: {self.X_test.shape[0]} samples")
        
        print(f"\nClass distribution:")
        train_dist = pd.Series(self.y_train).value_counts()
        val_dist = pd.Series(self.y_val).value_counts()
        test_dist = pd.Series(self.y_test).value_counts()
        
        print(f"    Training - LB: {train_dist.get(1, 0)}, NB: {train_dist.get(0, 0)}")
        print(f"    Validation - LB: {val_dist.get(1, 0)}, NB: {val_dist.get(0, 0)}")
        print(f"    Test - LB: {test_dist.get(1, 0)}, NB: {test_dist.get(0, 0)}")
        
        return True
    
    def perform_feature_selection_on_training_data(self):
        print("\n" + "="*50)
        print("STEP 3: FEATURE SELECTION ON TRAINING DATA ONLY")
        print("="*50)
        
        print("CRITICAL: Feature selection performed ONLY on training data")
        print("This prevents data leakage and ensures valid model evaluation")
        
        initial_features = self.X_train.shape[1]
        print(f"Initial number of features: {initial_features}")
        
        if self.raw_clinical_features is not None:
            n_clinical = self.raw_clinical_features.shape[1]
            clinical_feature_names = self.feature_names[:n_clinical]
            methylation_feature_names = self.feature_names[n_clinical:]
            
            X_train_clinical = self.X_train.iloc[:, :n_clinical]
            X_train_methylation = self.X_train.iloc[:, n_clinical:]
            
            print(f"    Clinical features in training: {n_clinical}")
            print(f"    Methylation features in training: {len(methylation_feature_names)}")
        else:
            clinical_feature_names = self.feature_names
            methylation_feature_names = []
            X_train_clinical = self.X_train
            X_train_methylation = pd.DataFrame()
        
        selected_methylation_features = []
        if len(methylation_feature_names) > 0:
            print("\n3.1 Methylation feature selection on training data")
            
            pvalues = []
            effect_sizes = []
            gene_names = []
            
            for i, gene in enumerate(methylation_feature_names):
                values = X_train_methylation.iloc[:, i].values
                
                # Skip if all values are the same, contain NaN, or have insufficient variance
                if len(np.unique(values)) == 1 or np.any(np.isnan(values)) or np.var(values) == 0:
                    continue
                
                # Split by B12 status (TRAINING DATA ONLY)
                lb_values = values[self.y_train == 1]
                nb_values = values[self.y_train == 0]
                
                if len(lb_values) == 0 or len(nb_values) == 0:
                    continue
                
                # Additional validation: check for sufficient variance in each group
                if np.var(lb_values) == 0 and np.var(nb_values) == 0:
                    continue
                
                try:
                    # Use Mann-Whitney U test for non-parametric comparison (more robust)
                    from scipy.stats import mannwhitneyu
                    statistic, pvalue = mannwhitneyu(lb_values, nb_values, alternative='two-sided')
                    
                    # Calculate Cohen's d (effect size) with error handling
                    mean_diff = np.mean(lb_values) - np.mean(nb_values)
                    pooled_var = ((len(lb_values) - 1) * np.var(lb_values, ddof=1) + 
                                (len(nb_values) - 1) * np.var(nb_values, ddof=1)) / (len(lb_values) + len(nb_values) - 2)
                    
                    if pooled_var > 0:
                        cohens_d = abs(mean_diff) / np.sqrt(pooled_var)
                    else:
                        cohens_d = 0
                    
                    # Additional check for valid results
                    if not (np.isnan(pvalue) or np.isnan(cohens_d) or np.isinf(pvalue) or np.isinf(cohens_d)):
                        pvalues.append(pvalue)
                        effect_sizes.append(cohens_d)
                        gene_names.append(gene)
                    
                except Exception as e:
                    # Skip problematic features
                    continue
            
            if len(gene_names) > 0:
                results_df = pd.DataFrame({
                    'Gene': gene_names,
                    'p_value': pvalues,
                    'effect_size': effect_sizes
                })
                
                print(f"    Total methylation genes tested: {len(results_df)}")
                print(f"    Significant genes (p < 0.05): {sum(results_df['p_value'] < 0.05)}")
                print(f"    Genes with large effect size (Cohen's d > 0.3): {sum(results_df['effect_size'] > 0.3)}")
                
                significant_genes = results_df[
                    (results_df['p_value'] < 0.05) & (results_df['effect_size'] > 0.3)
                ]['Gene'].tolist()
                
                if len(significant_genes) < 5:
                    print("    Too few features, relaxing criteria to p < 0.1")
                    significant_genes = results_df[
                        (results_df['p_value'] < 0.1) & (results_df['effect_size'] > 0.2)
                    ]['Gene'].tolist()
                
                if len(significant_genes) < 5:
                    print("    Still too few features, taking top 10 by effect size")
                    significant_genes = results_df.nlargest(10, 'effect_size')['Gene'].tolist()
                
                # For logistic regression, limit features more aggressively to prevent overfitting
                if len(significant_genes) > 10:
                    print("    Limiting to top 10 features for logistic regression stability with small sample")
                    top_genes = results_df.nlargest(10, 'effect_size')['Gene'].tolist()
                    significant_genes = top_genes
                
                selected_methylation_features = significant_genes
                print(f"    Selected methylation features: {len(selected_methylation_features)}")
                
                results_file = os.path.join(self.output_dir, f"{self.prefix}methylation_feature_selection_training_only_{self.timestamp}.csv")
                results_df.to_csv(results_file, index=False)
                print(f"    Feature selection results saved to: {results_file}")
        
        print(f"\n3.2 Combining selected features for Logistic Regression")
        selected_features = clinical_feature_names + selected_methylation_features
        print(f"    Total selected features: {len(selected_features)}")
        print(f"    Clinical features: {len(clinical_feature_names)}")
        print(f"    Methylation features: {len(selected_methylation_features)}")
        
        feature_indices = [self.feature_names.index(f) for f in selected_features]
        
        self.X_train = self.X_train.iloc[:, feature_indices]
        self.X_val = self.X_val.iloc[:, feature_indices]
        self.X_test = self.X_test.iloc[:, feature_indices]
        self.feature_names = selected_features
        
        print(f"\nFeature selection summary:")
        print(f"    Initial features: {initial_features}")
        print(f"    Final features: {len(self.feature_names)}")
        print(f"    Reduction: {initial_features - len(self.feature_names)} ({((initial_features - len(self.feature_names))/initial_features)*100:.1f}%)")
        
        feature_ratio = len(self.feature_names) / len(self.X_train)
        print(f"    Feature-to-sample ratio: {feature_ratio:.2f}")
        if feature_ratio > 0.1:
            print("    Warning: High feature-to-sample ratio may cause overfitting")
        else:
            print("    Good feature-to-sample ratio for stable logistic regression")
        
        return True
    
    def scale_features(self):
        print("\n" + "="*50)
        print("STEP 4: FEATURE SCALING (CRITICAL FOR LOGISTIC REGRESSION)")
        print("="*50)
        
        print("4.1 Data validation before scaling")
        
        # Check for NaN values before scaling
        nan_count_train = self.X_train.isnull().sum().sum()
        print(f"    NaN values in training data: {nan_count_train}")
        
        if nan_count_train > 0:
            print("    Handling NaN values with median imputation")
            from sklearn.impute import SimpleImputer
            imputer = SimpleImputer(strategy='median')
            
            # Fit imputer on training data only
            X_train_imputed = imputer.fit_transform(self.X_train)
            X_val_imputed = imputer.transform(self.X_val)
            X_test_imputed = imputer.transform(self.X_test)
            
            # Convert back to DataFrames
            self.X_train = pd.DataFrame(X_train_imputed, columns=self.feature_names, index=self.X_train.index)
            self.X_val = pd.DataFrame(X_val_imputed, columns=self.feature_names, index=self.X_val.index)
            self.X_test = pd.DataFrame(X_test_imputed, columns=self.feature_names, index=self.X_test.index)
            
            print(f"    NaN values after imputation: {self.X_train.isnull().sum().sum()}")
        
        # Check for infinite values
        inf_count = np.isinf(self.X_train.values).sum()
        print(f"    Infinite values in training data: {inf_count}")
        
        if inf_count > 0:
            print("    Replacing infinite values with finite values")
            self.X_train = self.X_train.replace([np.inf, -np.inf], np.nan)
            self.X_val = self.X_val.replace([np.inf, -np.inf], np.nan)
            self.X_test = self.X_test.replace([np.inf, -np.inf], np.nan)
            
            # Fill any remaining NaN values
            self.X_train = self.X_train.fillna(self.X_train.median())
            self.X_val = self.X_val.fillna(self.X_train.median())  # Use training median
            self.X_test = self.X_test.fillna(self.X_train.median())  # Use training median
        
        # Check for constant features (zero variance)
        feature_vars = self.X_train.var()
        constant_features = feature_vars[feature_vars == 0].index.tolist()
        print(f"    Constant features (zero variance): {len(constant_features)}")
        
        if len(constant_features) > 0:
            print(f"    Removing {len(constant_features)} constant features")
            remaining_features = [f for f in self.feature_names if f not in constant_features]
            
            self.X_train = self.X_train[remaining_features]
            self.X_val = self.X_val[remaining_features]
            self.X_test = self.X_test[remaining_features]
            self.feature_names = remaining_features
            
            print(f"    Features after constant removal: {len(self.feature_names)}")
        
        print("\n4.2 Applying StandardScaler to normalize feature scales")
        print("    IMPORTANT: Scaling fitted ONLY on training data")
        
        self.scaler = StandardScaler()
        
        X_train_scaled = self.scaler.fit_transform(self.X_train)
        print(f"    Scaler fitted on training data: {self.X_train.shape}")
        
        # Check for NaN values after scaling
        nan_after_scaling = np.isnan(X_train_scaled).sum()
        print(f"    NaN values after scaling: {nan_after_scaling}")
        
        if nan_after_scaling > 0:
            print("    Warning: NaN values found after scaling, replacing with 0")
            X_train_scaled = np.nan_to_num(X_train_scaled)
        
        X_val_scaled = self.scaler.transform(self.X_val)
        X_test_scaled = self.scaler.transform(self.X_test)
        
        # Replace any NaN values in validation and test sets
        X_val_scaled = np.nan_to_num(X_val_scaled)
        X_test_scaled = np.nan_to_num(X_test_scaled)
        
        self.X_train = pd.DataFrame(X_train_scaled, columns=self.feature_names, index=self.X_train.index)
        self.X_val = pd.DataFrame(X_val_scaled, columns=self.feature_names, index=self.X_val.index)
        self.X_test = pd.DataFrame(X_test_scaled, columns=self.feature_names, index=self.X_test.index)
        
        print(f"    Training data scaled: mean ~ 0, std ~ 1")
        print(f"    Validation data scaled using training parameters")
        print(f"    Test data scaled using training parameters")
        
        train_means = self.X_train.mean()
        train_stds = self.X_train.std()
        
        print(f"\n4.3 Scaling verification (training set):")
        print(f"    Feature means range: [{train_means.min():.6f}, {train_means.max():.6f}]")
        print(f"    Feature stds range: [{train_stds.min():.6f}, {train_stds.max():.6f}]")
        print(f"    Final NaN check: {self.X_train.isnull().sum().sum()} NaN values")
        
        print("    Feature scaling completed successfully")
        
        return True
    
    def train_logistic_regression(self):
        print("\n" + "="*50)
        print("STEP 5: LASSO LOGISTIC REGRESSION MODEL TRAINING")
        print("="*50)
        
        print("5.1 Training final LASSO model with L1 penalty")
        print("    Using the same L1 regularization as in stability selection")
        
        # Use LASSO (L1) regularization to be consistent with feature selection
        param_grid = {
            'C': [0.001, 0.01, 0.1, 1.0, 10.0, 100.0],  # Regularization strength
            'penalty': ['l1'],  # L1 penalty for sparsity (LASSO)
            'solver': ['liblinear'],  # Solver that supports L1 penalty
            'max_iter': [1000, 2000],  # Maximum iterations for convergence
        }
        
        print(f"    Parameter grid size: {len(param_grid['C']) * len(param_grid['penalty']) * len(param_grid['solver']) * len(param_grid['max_iter'])} combinations")
        
        # Initialize LASSO Logistic Regression
        lr_base = LogisticRegression(
            class_weight='balanced',  # Handle class imbalance
            random_state=42,
            n_jobs=1  # Single job for small dataset
        )
        
        # Perform grid search with cross-validation
        cv_strategy = StratifiedKFold(n_splits=min(5, len(self.X_train)//2), shuffle=True, random_state=42)
        
        print(f"    Using {cv_strategy.n_splits}-fold cross-validation")
        
        grid_search = GridSearchCV(
            lr_base,
            param_grid,
            cv=cv_strategy,
            scoring='roc_auc',
            n_jobs=1,
            verbose=1
        )
        
        print("    Starting LASSO hyperparameter optimization...")
        grid_search.fit(self.X_train, self.y_train)
        
        print(f"    Best cross-validation score: {grid_search.best_score_:.4f}")
        print(f"    Best parameters: {grid_search.best_params_}")
        
        self.model = grid_search.best_estimator_
        
        print(f"\n5.2 Final LASSO model configuration:")
        print(f"    Regularization (C): {self.model.C}")
        print(f"    Penalty: {self.model.penalty} (LASSO)")
        print(f"    Solver: {self.model.solver}")
        print(f"    Max iterations: {self.model.max_iter}")
        print(f"    Class weight: {self.model.class_weight}")
        
        # Analyze final model sparsity
        n_nonzero_coef = np.sum(np.abs(self.model.coef_[0]) > 1e-8)
        print(f"\n5.3 Model sparsity analysis:")
        print(f"    Total features: {len(self.feature_names)}")
        print(f"    Non-zero coefficients: {n_nonzero_coef}")
        print(f"    Sparsity level: {(len(self.feature_names) - n_nonzero_coef) / len(self.feature_names) * 100:.1f}%")
        
        if n_nonzero_coef < len(self.feature_names):
            print(f"    LASSO achieved additional sparsity: {len(self.feature_names) - n_nonzero_coef} features zeroed out")
        else:
            print(f"    All selected features retained (no additional sparsity from final LASSO)")
        
        print("    LASSO model training completed")
        
        return True
    
    def evaluate_model(self):
        print("\n" + "="*50)
        print("STEP 6: MODEL EVALUATION")
        print("="*50)
        
        y_pred = self.model.predict(self.X_test)
        y_pred_proba = self.model.predict_proba(self.X_test)[:, 1]
        
        print("6.1 Core Performance Metrics")
        
        auc_score = roc_auc_score(self.y_test, y_pred_proba)
        accuracy = accuracy_score(self.y_test, y_pred)
        f1 = f1_score(self.y_test, y_pred)
        
        print(f"    AUC (ROC): {auc_score:.4f}")
        print(f"    Accuracy: {accuracy:.4f}")
        print(f"    F1-Score: {f1:.4f}")
        
        cm = confusion_matrix(self.y_test, y_pred)
        print(f"\n6.2 Confusion Matrix:")
        print(cm)
        
        if cm.size == 4:
            tn, fp, fn, tp = cm.ravel()
            precision = tp / (tp + fp) if (tp + fp) > 0 else 0
            recall = tp / (tp + fn) if (tp + fn) > 0 else 0
            specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
            
            print(f"\n6.3 Clinical Performance Metrics")
            print(f"    Precision (PPV): {precision:.4f}")
            print(f"    Recall (Sensitivity): {recall:.4f}")
            print(f"    Specificity: {specificity:.4f}")
        
        print("\n6.4 Detailed Classification Report")
        class_names = self.label_encoder.classes_
        print(classification_report(self.y_test, y_pred, target_names=class_names))
        
        cv_scores = cross_val_score(self.model, self.X_train, self.y_train, 
                                   cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
                                   scoring='roc_auc')
        
        print(f"\n6.5 Cross-Validation Assessment (Training Data)")
        print(f"    Cross-validation AUC scores: {cv_scores}")
        print(f"    Mean CV AUC: {cv_scores.mean():.4f} (+/-{cv_scores.std()*2:.4f})")
        
        evaluation_results = {
            'test_metrics': {
                'auc_roc': auc_score,
                'accuracy': accuracy,
                'f1_score': f1,
                'precision': precision if cm.size == 4 else 0,
                'recall_sensitivity': recall if cm.size == 4 else 0,
                'specificity': specificity if cm.size == 4 else 0
            },
            'confusion_matrix': cm.tolist(),
            'cross_validation': {
                'cv_scores': cv_scores.tolist(),
                'mean_cv_auc': cv_scores.mean(),
                'std_cv_auc': cv_scores.std()
            },
            'predictions': {
                'y_test': self.y_test.tolist(),
                'y_pred': y_pred.tolist(),
                'y_pred_proba': y_pred_proba.tolist()
            }
        }
        
        evaluation_file = os.path.join(self.output_dir, f"{self.prefix}evaluation_results_{self.timestamp}.json")
        with open(evaluation_file, 'w') as f:
            json.dump(evaluation_results, f, indent=4)
        
        print(f"    Evaluation results saved to: {evaluation_file}")
        
        return evaluation_results
    
    def analyze_feature_coefficients(self):
        print("\n" + "="*50)
        print("STEP 7: LOGISTIC REGRESSION COEFFICIENTS ANALYSIS")
        print("="*50)
        
        coefficients = self.model.coef_[0]
        
        feature_coef_df = pd.DataFrame({
            'feature': self.feature_names,
            'coefficient': coefficients,
            'abs_coefficient': np.abs(coefficients)
        }).sort_values('abs_coefficient', ascending=False)
        
        print("7.1 Top 15 Most Important Features by Coefficient Magnitude:")
        top_features = feature_coef_df.head(15)
        for idx, row in top_features.iterrows():
            direction = "increases" if row['coefficient'] > 0 else "decreases"
            print(f"    {row['feature']}: {row['coefficient']:.4f} ({direction} LB probability)")
        
        clinical_features = ['age', 'BMI', 'parity', 'B12supplem', 
                           'v1p_MultivitTab', 'v1p_FolicAcid', 'smoking']
        
        clinical_coef = feature_coef_df[
            feature_coef_df['feature'].isin(clinical_features)
        ]
        methylation_coef = feature_coef_df[
            ~feature_coef_df['feature'].isin(clinical_features)
        ]
        
        print(f"\n7.2 Feature Type Analysis:")
        print(f"    Clinical features: {len(clinical_coef)}")
        print(f"    Methylation features: {len(methylation_coef)}")
        
        if len(clinical_coef) > 0:
            clinical_total_importance = clinical_coef['abs_coefficient'].sum()
            print(f"    Total clinical coefficient magnitude: {clinical_total_importance:.4f}")
            print(f"    Average clinical coefficient magnitude: {clinical_coef['abs_coefficient'].mean():.4f}")
            print(f"    Top clinical feature: {clinical_coef.iloc[0]['feature']} ({clinical_coef.iloc[0]['coefficient']:.4f})")
        
        if len(methylation_coef) > 0:
            methylation_total_importance = methylation_coef['abs_coefficient'].sum()
            print(f"    Total methylation coefficient magnitude: {methylation_total_importance:.4f}")
            print(f"    Average methylation coefficient magnitude: {methylation_coef['abs_coefficient'].mean():.4f}")
            print(f"    Top methylation feature: {methylation_coef.iloc[0]['feature']} ({methylation_coef.iloc[0]['coefficient']:.4f})")
        
        total_importance = feature_coef_df['abs_coefficient'].sum()
        if len(clinical_coef) > 0 and len(methylation_coef) > 0:
            clinical_pct = (clinical_coef['abs_coefficient'].sum() / total_importance) * 100
            methylation_pct = (methylation_coef['abs_coefficient'].sum() / total_importance) * 100
            print(f"\n7.3 Feature Type Contribution:")
            print(f"    Clinical features contribute: {clinical_pct:.1f}% of total coefficient magnitude")
            print(f"    Methylation features contribute: {methylation_pct:.1f}% of total coefficient magnitude")
        
        coef_file = os.path.join(self.output_dir, f"{self.prefix}feature_coefficients_{self.timestamp}.csv")
        feature_coef_df.to_csv(coef_file, index=False)
        print(f"    Feature coefficients saved to: {coef_file}")
        
        return feature_coef_df
    
    def create_visualizations(self, evaluation_results, feature_coef_df):
        print("\n" + "="*50)
        print("STEP 8: CREATING VISUALIZATIONS")
        print("="*50)
        
        plt.style.use('default')
        
        fig = plt.figure(figsize=(16, 12))
        
        print("8.1 Creating ROC curve")
        plt.subplot(2, 3, 1)
        y_pred_proba = evaluation_results['predictions']['y_pred_proba']
        y_test = evaluation_results['predictions']['y_test']
        
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        auc_score = evaluation_results['test_metrics']['auc_roc']
        
        plt.plot(fpr, tpr, linewidth=2, label=f'ROC Curve (AUC = {auc_score:.3f})')
        plt.plot([0, 1], [0, 1], 'k--', linewidth=1)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve - Logistic Regression')
        plt.legend(loc="lower right")
        plt.grid(True, alpha=0.3)
        
        print("8.2 Creating Precision-Recall curve")
        plt.subplot(2, 3, 2)
        
        precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
        f1_score_val = evaluation_results['test_metrics']['f1_score']
        
        plt.plot(recall, precision, linewidth=2, label=f'PR Curve (F1 = {f1_score_val:.3f})')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision-Recall Curve')
        plt.legend(loc="lower left")
        plt.grid(True, alpha=0.3)
        
        print("8.3 Creating confusion matrix heatmap")
        plt.subplot(2, 3, 3)
        
        cm = np.array(evaluation_results['confusion_matrix'])
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=['NB', 'LB'], yticklabels=['NB', 'LB'])
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        
        print("8.4 Creating coefficient magnitude plot")
        plt.subplot(2, 3, 4)
        
        top_15_features = feature_coef_df.head(15)
        clinical_features = ['age', 'BMI', 'parity', 'B12supplem', 'v1p_MultivitTab', 'v1p_FolicAcid', 'smoking']
        colors = ['blue' if f in clinical_features else 'green' for f in top_15_features['feature']]
        
        plt.barh(range(len(top_15_features)), top_15_features['abs_coefficient'], color=colors)
        plt.yticks(range(len(top_15_features)), 
                  [f[:25] + '...' if len(f) > 25 else f for f in top_15_features['feature']])
        plt.xlabel('Coefficient Magnitude')
        plt.title('Top 15 Features (Blue=Clinical, Green=Methylation)')
        plt.gca().invert_yaxis()
        
        print("8.5 Creating coefficient direction plot")
        plt.subplot(2, 3, 5)
        
        positive_coef = feature_coef_df[feature_coef_df['coefficient'] > 0].head(10)
        negative_coef = feature_coef_df[feature_coef_df['coefficient'] < 0].head(10)
        
        plt.barh(range(len(positive_coef)), positive_coef['coefficient'], 
                color='red', alpha=0.7, label='Increases LB Risk')
        plt.barh(range(len(positive_coef), len(positive_coef) + len(negative_coef)), 
                negative_coef['coefficient'], color='blue', alpha=0.7, label='Decreases LB Risk')
        
        all_labels = list(positive_coef['feature']) + list(negative_coef['feature'])
        plt.yticks(range(len(all_labels)), 
                  [f[:20] + '...' if len(f) > 20 else f for f in all_labels])
        plt.xlabel('Coefficient Value')
        plt.title('Feature Effects on B12 Deficiency')
        plt.legend()
        plt.gca().invert_yaxis()
        
        print("8.6 Creating cross-validation scores plot")
        plt.subplot(2, 3, 6)
        
        cv_scores = evaluation_results['cross_validation']['cv_scores']
        plt.bar(range(1, len(cv_scores)+1), cv_scores, color='green', alpha=0.7)
        plt.axhline(y=np.mean(cv_scores), color='red', linestyle='--', 
                   label=f'Mean: {np.mean(cv_scores):.3f}')
        plt.xlabel('CV Fold')
        plt.ylabel('AUC Score')
        plt.title('Cross-Validation Scores')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        
        plot_file = os.path.join(self.output_dir, f"{self.prefix}model_evaluation_plots_{self.timestamp}.png")
        plt.savefig(plot_file, dpi=300, bbox_inches='tight')
        print(f"    Visualization plots saved to: {plot_file}")
        
        plt.show()
        
        return True
    
    def save_model(self):
        print("\n" + "="*50)
        print("STEP 9: SAVING MODEL")
        print("="*50)
        
        model_file = os.path.join(self.output_dir, f"{self.prefix}trained_model_{self.timestamp}.joblib")
        joblib.dump(self.model, model_file)
        print(f"    Trained model saved to: {model_file}")
        
        scaler_file = os.path.join(self.output_dir, f"{self.prefix}scaler_{self.timestamp}.joblib")
        joblib.dump(self.scaler, scaler_file)
        print(f"    Feature scaler saved to: {scaler_file}")
        
        encoder_file = os.path.join(self.output_dir, f"{self.prefix}label_encoder_{self.timestamp}.joblib")
        joblib.dump(self.label_encoder, encoder_file)
        print(f"    Label encoder saved to: {encoder_file}")
        
        features_file = os.path.join(self.output_dir, f"{self.prefix}feature_names_{self.timestamp}.json")
        with open(features_file, 'w') as f:
            json.dump(self.feature_names, f, indent=4)
        print(f"    Feature names saved to: {features_file}")
        
        return True
    
    def run_complete_pipeline(self, clinical_path, dmr_path, methylation_path):
        print("\n" + "="*80)
        print("STARTING COMPLETE B12 PREDICTION MODEL PIPELINE - LOGISTIC REGRESSION")
        print("NO DATA LEAKAGE VERSION")
        print("="*80)
        
        pipeline_start_time = datetime.now()
        
        try:
            # Step 1: Load and preprocess data
            if not self.load_data(clinical_path, dmr_path, methylation_path):
                print("ERROR: Data loading failed")
                return False
            
            if not self.prepare_initial_dataset():
                print("ERROR: Initial dataset preparation failed")
                return False
            
            # Step 2: Split data BEFORE feature selection
            if not self.split_data():
                print("ERROR: Data splitting failed")
                return False
            
            # Step 3: Feature selection with stability selection on training data only
            if not self.perform_feature_selection_on_training_data():
                print("ERROR: Feature selection failed")
                return False
            
            # Step 4: Scale features
            if not self.scale_features():
                print("ERROR: Feature scaling failed")
                return False
            
            # Step 5: Train model
            if not self.train_logistic_regression():
                print("ERROR: Model training failed")
                return False
            
            # Step 6: Evaluate model
            evaluation_results = self.evaluate_model()
            if not evaluation_results:
                print("ERROR: Model evaluation failed")
                return False
            
            # Step 7: Analyze coefficients
            feature_coef_df = self.analyze_feature_coefficients()
            
            # Step 8: Create visualizations
            self.create_visualizations(evaluation_results, feature_coef_df)
            
            # Step 9: Save model
            self.save_model()
            
            # Pipeline completion summary
            pipeline_end_time = datetime.now()
            pipeline_duration = pipeline_end_time - pipeline_start_time
            
            print("\n" + "="*80)
            print("PIPELINE COMPLETION SUMMARY - LOGISTIC REGRESSION")
            print("="*80)
            print(f"Start time: {pipeline_start_time.strftime('%Y-%m-%d %H:%M:%S')}")
            print(f"End time: {pipeline_end_time.strftime('%Y-%m-%d %H:%M:%S')}")
            print(f"Total duration: {pipeline_duration}")
            
            print("\n" + "="*60)
            print("KEY MODEL PERFORMANCE METRICS")
            print("="*60)
            
            auc = evaluation_results['test_metrics']['auc_roc']
            accuracy = evaluation_results['test_metrics']['accuracy']
            precision = evaluation_results['test_metrics']['precision']
            recall = evaluation_results['test_metrics']['recall_sensitivity']
            
            print(f"1. AUC (Area Under ROC Curve):     {auc:.4f}")
            print(f"   Model's ability to distinguish between B12 deficient and normal")
            print(f"2. Accuracy (Overall Correctness):  {accuracy:.4f}")
            print(f"   Proportion of correct predictions (both LB and NB)")
            print(f"3. Precision (Positive Predictive): {precision:.4f}")
            print(f"   Among predicted deficient, how many are truly deficient")
            print(f"4. Recall (Sensitivity):            {recall:.4f}")
            print(f"   Among truly deficient, how many were correctly identified")
            print("="*60)
            
            print("\nMODEL PERFORMANCE INTERPRETATION:")
            if auc >= 0.9:
                print("• AUC: Excellent discrimination ability")
            elif auc >= 0.8:
                print("• AUC: Good discrimination ability")
            elif auc >= 0.7:
                print("• AUC: Fair discrimination ability")
            else:
                print("• AUC: Poor discrimination ability")
                
            if accuracy >= 0.9:
                print("• Accuracy: Excellent overall performance")
            elif accuracy >= 0.8:
                print("• Accuracy: Good overall performance")
            elif accuracy >= 0.7:
                print("• Accuracy: Moderate overall performance")
            else:
                print("• Accuracy: Poor overall performance")
                
            if precision >= 0.8:
                print("• Precision: Low false positive rate - reliable when predicting deficiency")
            elif precision >= 0.6:
                print("• Precision: Moderate false positive rate")
            else:
                print("• Precision: High false positive rate - many false alarms")
                
            if recall >= 0.8:
                print("• Recall: Low false negative rate - catches most deficient cases")
            elif recall >= 0.6:
                print("• Recall: Moderate false negative rate")
            else:
                print("• Recall: High false negative rate - misses many deficient cases")
            
            print(f"\nNumber of features used: {len(self.feature_names)}")
            print(f"Test set size: {len(self.y_test)} samples")
            print(f"Output directory: {self.output_dir}")
            
            clinical_features = ['age', 'BMI', 'parity', 'B12supplem', 'v1p_MultivitTab', 'v1p_FolicAcid', 'smoking']
            n_clinical = sum(1 for f in self.feature_names if f in clinical_features)
            n_methylation = len(self.feature_names) - n_clinical
            
            print(f"\nFeature composition:")
            print(f"    Clinical features: {n_clinical}")
            print(f"    Methylation features: {n_methylation}")
            
            if n_methylation > n_clinical:
                print("Good: Model is primarily driven by methylation features")
            elif n_clinical > n_methylation:
                print("Warning: Model is primarily driven by clinical features")
            else:
                print("Balanced: Equal clinical and methylation features")
            
            print("="*80)
            
            summary_report = {
                'pipeline_info': {
                    'model_type': 'Logistic Regression',
                    'version': 'No Data Leakage',
                    'start_time': pipeline_start_time.isoformat(),
                    'end_time': pipeline_end_time.isoformat(),
                    'duration_seconds': pipeline_duration.total_seconds(),
                    'success': True,
                    'data_leakage_prevented': True,
                    'b12_mol_excluded': True,
                    'feature_scaling_applied': True
                },
                'key_metrics': {
                    'AUC': auc,
                    'Accuracy': accuracy,
                    'Precision': precision,
                    'Recall_Sensitivity': recall
                },
                'data_info': {
                    'total_samples': len(self.X_train) + len(self.X_val) + len(self.X_test),
                    'total_features': len(self.feature_names),
                    'train_samples': len(self.X_train),
                    'val_samples': len(self.X_val),
                    'test_samples': len(self.X_test),
                    'clinical_features': n_clinical,
                    'methylation_features': n_methylation
                },
                'model_performance': evaluation_results['test_metrics'],
                'top_features': feature_coef_df.head(10).to_dict('records'),
                'feature_selection_method': 'TRUE Stability Selection with LASSO L1 penalty on training data only',
                'model_parameters': {
                    'C': self.model.C,
                    'penalty': self.model.penalty,
                    'solver': self.model.solver,
                    'class_weight': str(self.model.class_weight)
                }
            }
            
            summary_file = os.path.join(self.output_dir, f"{self.prefix}final_summary_report_no_leakage_{self.timestamp}.json")
            with open(summary_file, 'w') as f:
                json.dump(summary_report, f, indent=4)
            
            print(f"Final summary report saved to: {summary_file}")
            
            return True
            
        except Exception as e:
            print(f"ERROR: Pipeline failed with exception: {str(e)}")
            self.logger.error(f"Pipeline failed with exception: {str(e)}")
            import traceback
            traceback.print_exc()
            return False


def main():
    """Main function to run the B12 prediction model - Logistic Regression"""
    
    # File paths
    clinical_path = "/Users/heweilin/Desktop/P056/7Clinical_data50.csv"
    dmr_path = "/Users/heweilin/Desktop/P056/4DNA_DMRs.csv"
    methylation_path = "/Users/heweilin/Desktop/P056_Code_2/Processed_Data/1_PD_PromoterRegion_CpGs.csv"
    output_dir = "/Users/heweilin/Desktop/P056_Code_3/Data"
    
    print("B12 Deficiency Prediction Model - Stage 1 Logistic Regression")
    print("FIXED DATA LEAKAGE VERSION")
    print("=" * 60)
    print("Key Improvements:")
    print("- Excluded B12_mol to prevent target leakage")
    print("- Feature selection performed ONLY on training data")
    print("- Feature scaling applied for logistic regression")
    print("- Larger test set (25% instead of 15%)")
    print("- Focus on methylation features over clinical biomarkers")
    print("- Reproducible results with fixed random seeds")
    print("=" * 60)
    print(f"Clinical data: {clinical_path}")
    print(f"DMR data: {dmr_path}")
    print(f"Methylation data: {methylation_path}")
    print(f"Output directory: {output_dir}")
    print("=" * 60)
    
    # Initialize the model class
    model = B12PredictionLR(output_dir=output_dir, prefix="LR_Stage1_NoLeakage_")
    
    # Run the complete pipeline
    success = model.run_complete_pipeline(clinical_path, dmr_path, methylation_path)
    
    if success:
        print("\n" + "="*80)
        print("MODEL PIPELINE COMPLETED SUCCESSFULLY!")
        print("="*80)
        print("Key Performance Achieved:")
        
        try:
            results_file = os.path.join(output_dir, f"LR_Stage1_NoLeakage_evaluation_results_{model.timestamp}.json")
            if os.path.exists(results_file):
                with open(results_file, 'r') as f:
                    results = json.load(f)
                
                print(f"• AUC (Discrimination):    {results['test_metrics']['auc_roc']:.4f}")
                print(f"• Accuracy (Correctness):  {results['test_metrics']['accuracy']:.4f}")
                print(f"• Precision (PPV):         {results['test_metrics']['precision']:.4f}")
                print(f"• Recall (Sensitivity):    {results['test_metrics']['recall_sensitivity']:.4f}")
        except:
            print("• Check output files for detailed metrics")
        
        print("\nLogistic Regression Model Features:")
        print("• Interpretable coefficients showing feature effects")
        print("• Probabilistic predictions for risk assessment")
        print("• L2 regularization prevents overfitting")
        print("• Feature scaling ensures stable convergence")
        print("• No data leakage issues")
        print("Check the output directory for all generated files.")
        print("="*80)
    else:
        print("\nMODEL PIPELINE FAILED!")
        print("Check the log files for detailed error information.")
    
    return success


if __name__ == "__main__":
    # Set random seeds for reproducibility
    np.random.seed(42)
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
B12 Deficiency Prediction Model - Data-Driven Feature Selection
Multiple Train-Test Splits for Robust Feature Selection

Author: Research Team
Date: August 25, 2025
Version: 3.0 - Multiple Splits Approach with Cleaned Code

Strategy: Multiple independent train-test splits to identify stable features
Objective: Predict B12 deficiency using robust, consistently selected features
"""

import os
import sys
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import (train_test_split, StratifiedKFold, 
                                   cross_val_score, GridSearchCV)
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (classification_report, confusion_matrix, 
                           roc_auc_score, roc_curve, precision_recall_curve,
                           f1_score, accuracy_score)
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json
from datetime import datetime
import logging

class B12PredictionDataDriven:
    """B12 deficiency prediction with multiple splits approach"""
    
    def __init__(self, output_dir="/Users/heweilin/Desktop/P056_Code_3/Data", prefix="DataDriven_"):
        self.output_dir = output_dir
        self.prefix = prefix
        self.timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        
        os.makedirs(output_dir, exist_ok=True)
        self.setup_logging()
        
        # Initialize data storage
        self.clinical_data = None
        self.methylation_data = None
        self.dmr_data = None
        self.X = None
        self.y = None
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        self.feature_names = None
        self.model = None
        self.preprocessor = None
        self.label_encoder = None
        self.selected_methylation_genes = []
        self.methylation_genes = []
        
        print("=" * 80)
        print("B12 DEFICIENCY PREDICTION - MULTIPLE SPLITS APPROACH")
        print("=" * 80)
        print(f"Timestamp: {self.timestamp}")
        print(f"Output directory: {output_dir}")
        print("Key Features:")
        print("- Multiple train-test splits for robust feature selection")
        print("- Batch correction for methylation data")
        print("- Mann-Whitney U + Cohen's d feature selection")
        print("- L1 regularization (LASSO) logistic regression")
        print("=" * 80)
    
    def setup_logging(self):
        log_file = os.path.join(self.output_dir, f"{self.prefix}model_log_{self.timestamp}.txt")
        logging.basicConfig(
            level=logging.INFO,
            format='%(asctime)s - %(levelname)s - %(message)s',
            handlers=[
                logging.FileHandler(log_file),
                logging.StreamHandler(sys.stdout)
            ]
        )
        self.logger = logging.getLogger(__name__)
        self.logger.info("Data-driven model logging setup completed")
    
    def load_data(self, clinical_path, dmr_path, methylation_path):
        print("\n" + "=" * 50)
        print("STEP 1: DATA LOADING")
        print("=" * 50)
        
        print(f"Loading clinical data from: {clinical_path}")
        try:
            self.clinical_data = pd.read_csv(clinical_path)
            print(f"Clinical data shape: {self.clinical_data.shape}")
            self.logger.info(f"Clinical data loaded: {self.clinical_data.shape}")
        except Exception as e:
            print(f"Error loading clinical data: {e}")
            return False
        
        print(f"Loading DMR data from: {dmr_path}")
        try:
            self.dmr_data = pd.read_csv(dmr_path, low_memory=False)
            print(f"DMR data shape: {self.dmr_data.shape}")
            self.logger.info(f"DMR data loaded: {self.dmr_data.shape}")
        except Exception as e:
            print(f"Error loading DMR data: {e}")
            return False
        
        print(f"Loading methylation data from: {methylation_path}")
        try:
            self.methylation_data = pd.read_csv(methylation_path)
            print(f"Methylation data shape: {self.methylation_data.shape}")
            self.logger.info(f"Methylation data loaded: {self.methylation_data.shape}")
        except Exception as e:
            print(f"Error loading methylation data: {e}")
            return False
        
        return True
    
    def prepare_methylation_data_with_batch_correction(self):
        print("\n" + "=" * 50)
        print("STEP 2: BATCH CORRECTION")
        print("=" * 50)
        
        methylation_df = self.methylation_data.set_index('Gene_Symbol').T
        print(f"Methylation data after transpose: {methylation_df.shape}")
        
        clinical_ids = self.clinical_data['DNA_ID'].values
        methylation_samples = methylation_df.index.values
        
        common_samples = list(set(clinical_ids) & set(methylation_samples))
        print(f"Common samples: {len(common_samples)}")
        
        if len(common_samples) == 0:
            print("Warning: No common samples found")
            return None, None
        
        methylation_filtered = methylation_df.loc[common_samples]
        clinical_aligned = self.clinical_data[self.clinical_data['DNA_ID'].isin(common_samples)]
        clinical_aligned = clinical_aligned.set_index('DNA_ID').loc[common_samples]
        
        if 'batch_DNA' in clinical_aligned.columns:
            batch_info = clinical_aligned['batch_DNA'].values
            unique_batches = np.unique(batch_info[~pd.isna(batch_info)])
            print(f"Found DNA batches: {unique_batches}")
            print(f"Batch distribution: {pd.Series(batch_info).value_counts().to_dict()}")
            
            print("Performing batch correction...")
            methylation_corrected = self.apply_batch_correction(methylation_filtered, batch_info)
            print(f"Batch-corrected data shape: {methylation_corrected.shape}")
        else:
            print("No batch information found, skipping correction")
            methylation_corrected = methylation_filtered
        
        return methylation_corrected, common_samples
    
    def apply_batch_correction(self, methylation_data, batch_info):
        corrected_data = methylation_data.copy()
        batches = pd.Categorical(batch_info)
        unique_batches = batches.categories
        
        if len(unique_batches) <= 1:
            print("Only one batch found, no correction needed")
            return corrected_data
        
        print(f"Correcting for {len(unique_batches)} batches")
        n_genes_corrected = 0
        
        for gene in corrected_data.columns:
            gene_values = corrected_data[gene].values
            
            if np.sum(~np.isnan(gene_values)) < len(gene_values) * 0.5:
                continue
            if np.nanvar(gene_values) == 0:
                continue
            
            try:
                corrected_values = self.correct_gene_for_batch(gene_values, batches)
                corrected_data.loc[:, gene] = corrected_values
                n_genes_corrected += 1
            except Exception:
                continue
        
        print(f"Successfully corrected {n_genes_corrected} genes for batch effects")
        return corrected_data
    
    def correct_gene_for_batch(self, gene_values, batches):
        df = pd.DataFrame({'values': gene_values, 'batch': batches})
        df_clean = df.dropna()
        
        if len(df_clean) < 3:
            return gene_values
        
        overall_mean = df_clean['values'].mean()
        batch_means = df_clean.groupby('batch')['values'].mean()
        batch_effects = batch_means - overall_mean
        
        corrected_values = gene_values.copy()
        for batch_name, batch_effect in batch_effects.items():
            batch_mask = (batches == batch_name)
            corrected_values[batch_mask] = corrected_values[batch_mask] - batch_effect
        
        return corrected_values
    
    def prepare_combined_dataset(self):
        print("\n" + "=" * 50)
        print("STEP 3: COMBINE CLINICAL AND METHYLATION DATA")
        print("=" * 50)
        
        # Clinical features (excluding B12_mol and batch_DNA)
        numerical_features = ['age', 'BMI', 'parity', 'B12supplem', 
                             'v1p_MultivitTab', 'v1p_FolicAcid', 'smoking']
        categorical_features = ['ethnicity', 'v3n_Gender', 'BMI_cat']
        
        available_numerical = [f for f in numerical_features if f in self.clinical_data.columns]
        available_categorical = [f for f in categorical_features if f in self.clinical_data.columns]
        
        print(f"Available numerical features: {available_numerical}")
        print(f"Available categorical features: {available_categorical}")
        print("EXCLUDED: B12_mol (target leakage), batch_DNA (used for correction)")
        
        all_clinical_features = available_numerical + available_categorical
        clinical_X = self.clinical_data[all_clinical_features].copy()
        
        methylation_result = self.prepare_methylation_data_with_batch_correction()
        if methylation_result[0] is None:
            print("Using only clinical features")
            self.X = clinical_X
            self.feature_names = list(clinical_X.columns)
            self.y = self.clinical_data['B12_status'].values
            self.methylation_genes = []
        else:
            methylation_X, common_samples = methylation_result
            print("Combining clinical and methylation features")
            
            clinical_filtered = self.clinical_data[self.clinical_data['DNA_ID'].isin(common_samples)]
            clinical_filtered = clinical_filtered.set_index('DNA_ID').loc[common_samples]
            clinical_X_filtered = clinical_filtered[all_clinical_features]
            
            methylation_X_aligned = methylation_X.loc[common_samples]
            
            self.X = pd.concat([clinical_X_filtered, methylation_X_aligned], axis=1)
            self.feature_names = list(clinical_X_filtered.columns) + list(methylation_X_aligned.columns)
            self.methylation_genes = list(methylation_X_aligned.columns)
            
            self.y = clinical_filtered['B12_status'].values
        
        print(f"Combined dataset shape: {self.X.shape}")
        print(f"Clinical features: {len(self.feature_names) - len(self.methylation_genes)}")
        print(f"Methylation features: {len(self.methylation_genes)}")
        
        self.label_encoder = LabelEncoder()
        self.y = self.label_encoder.fit_transform(self.y)
        print(f"Target encoding: {dict(zip(self.label_encoder.classes_, self.label_encoder.transform(self.label_encoder.classes_)))}")
        
        return True
    
    def run_multiple_train_test_splits(self, n_iterations=20, test_size=0.2):
        print("\n" + "=" * 50)
        print("STEP 4: MULTIPLE TRAIN-TEST SPLITS ANALYSIS")
        print("=" * 50)
        
        print(f"Running {n_iterations} independent train-test splits")
        print(f"Each split: {(1-test_size)*100:.0f}% training, {test_size*100:.0f}% testing")
        
        all_selected_features = []
        all_performances = []
        iteration_details = []
        
        for iteration in range(n_iterations):
            print(f"\nIteration {iteration + 1}/{n_iterations}")
            
            X_iter_train, X_iter_test, y_iter_train, y_iter_test = train_test_split(
                self.X, self.y, test_size=test_size, 
                random_state=42 + iteration,
                stratify=self.y
            )
            
            print(f"Train: {len(X_iter_train)} samples, Test: {len(X_iter_test)} samples")
            
            try:
                selected_features = self.select_features_for_iteration(X_iter_train, y_iter_train)
                
                if len(selected_features) == 0:
                    print(f"Warning: No features selected in iteration {iteration + 1}")
                    continue
                
                performance = self.train_evaluate_iteration(
                    X_iter_train, X_iter_test, y_iter_train, y_iter_test, selected_features
                )
                
                all_selected_features.append(selected_features)
                all_performances.append(performance)
                
                iteration_details.append({
                    'iteration': iteration + 1,
                    'selected_features': selected_features,
                    'performance': performance,
                    'n_train': len(X_iter_train),
                    'n_test': len(X_iter_test)
                })
                
                print(f"Selected {len(selected_features)} features, AUC: {performance['auc']:.3f}")
                
            except Exception as e:
                print(f"Error in iteration {iteration + 1}: {e}")
                continue
        
        # Feature stability analysis
        print(f"\nFEATURE STABILITY ANALYSIS ({len(all_selected_features)} successful iterations)")
        print("=" * 60)
        
        feature_counts = {}
        for selected_features in all_selected_features:
            for feature in selected_features:
                feature_counts[feature] = feature_counts.get(feature, 0) + 1
        
        stable_features = sorted(feature_counts.items(), key=lambda x: x[1], reverse=True)
        
        print("Most consistently selected features:")
        for i, (feature, count) in enumerate(stable_features[:15], 1):
            stability_pct = (count / len(all_selected_features)) * 100
            stability_icon = "★" if stability_pct >= 70 else "◆" if stability_pct >= 50 else "○"
            print(f"  {i:2d}. {stability_icon} {feature:<15} | {count}/{len(all_selected_features)} iterations ({stability_pct:.0f}%)")
        
        # Performance statistics
        valid_performances = [p for p in all_performances if p is not None]
        if valid_performances:
            aucs = [p['auc'] for p in valid_performances]
            accuracies = [p['accuracy'] for p in valid_performances]
            
            print(f"\nPERFORMACE STABILITY:")
            print(f"  Mean AUC: {np.mean(aucs):.3f} ± {np.std(aucs):.3f}")
            print(f"  AUC Range: [{np.min(aucs):.3f}, {np.max(aucs):.3f}]")
            print(f"  Mean Accuracy: {np.mean(accuracies):.3f} ± {np.std(accuracies):.3f}")
            print(f"  Iterations with AUC > 0.8: {sum(1 for auc in aucs if auc > 0.8)}/{len(aucs)}")
        
        # Feature stability insights
        highly_stable = sum(1 for _, count in stable_features if count >= len(all_selected_features) * 0.7)
        moderately_stable = sum(1 for _, count in stable_features if count >= len(all_selected_features) * 0.5)
        
        print(f"\nSTABILITY INSIGHTS:")
        print(f"  Features selected ≥70% of iterations: {highly_stable}")
        print(f"  Features selected ≥50% of iterations: {moderately_stable}")
        print(f"  Total unique features selected: {len(stable_features)}")
        
        if highly_stable >= 5:
            print(f"  ✓ Good stability: {highly_stable} highly stable features found")
        elif moderately_stable >= 5:
            print(f"  ~ Moderate stability: {moderately_stable} moderately stable features")
        else:
            print(f"  ✗ Poor stability: High feature variability across iterations")
        
        results_summary = {
            'n_iterations': n_iterations,
            'successful_iterations': len(all_selected_features),
            'feature_stability': feature_counts,
            'stable_features_ranking': stable_features,
            'performance_stats': {
                'mean_auc': np.mean(aucs) if valid_performances else None,
                'std_auc': np.std(aucs) if valid_performances else None,
                'mean_accuracy': np.mean(accuracies) if valid_performances else None,
                'auc_range': [np.min(aucs), np.max(aucs)] if valid_performances else None
            },
            'iteration_details': iteration_details,
            'stability_insights': {
                'highly_stable_features': highly_stable,
                'moderately_stable_features': moderately_stable,
                'total_unique_features': len(stable_features)
            }
        }
        
        results_file = os.path.join(self.output_dir, f"{self.prefix}multiple_splits_analysis_{self.timestamp}.json")
        with open(results_file, 'w') as f:
            json.dump(results_summary, f, indent=4)
        print(f"\nComplete analysis saved to: {results_file}")
        
        return results_summary
    
    def select_features_for_iteration(self, X_train, y_train):
        clinical_features = [f for f in self.feature_names if f not in self.methylation_genes]
        methylation_features = [col for col in X_train.columns if col in self.methylation_genes]
        
        feature_stats = []
        
        for gene in methylation_features:
            values = X_train[gene].values
            values = pd.to_numeric(values, errors='coerce')
            
            if len(np.unique(values[~np.isnan(values)])) <= 1:
                continue
            if np.sum(~np.isnan(values)) < len(values) * 0.5:
                continue
            if np.nanvar(values) == 0:
                continue
            
            valid_mask = ~np.isnan(values)
            if np.sum(valid_mask) < 6:
                continue
            
            values_clean = values[valid_mask]
            y_clean = y_train[valid_mask]
            
            lb_values = values_clean[y_clean == 1]
            nb_values = values_clean[y_clean == 0]
            
            if len(lb_values) < 2 or len(nb_values) < 2:
                continue
            
            try:
                statistic, pvalue = mannwhitneyu(lb_values, nb_values, alternative='two-sided')
                
                mean_diff = np.mean(lb_values) - np.mean(nb_values)
                pooled_var = ((len(lb_values) - 1) * np.var(lb_values, ddof=1) + 
                            (len(nb_values) - 1) * np.var(nb_values, ddof=1)) / (len(lb_values) + len(nb_values) - 2)
                
                cohens_d = abs(mean_diff) / np.sqrt(pooled_var) if pooled_var > 0 else 0
                
                if not (np.isnan(pvalue) or np.isnan(cohens_d)):
                    feature_stats.append({
                        'Gene': gene,
                        'p_value': pvalue,
                        'effect_size': cohens_d,
                        'composite_score': cohens_d * (-np.log10(pvalue + 1e-10))
                    })
                    
            except Exception:
                continue
        
        if len(feature_stats) == 0:
            return clinical_features
        
        results_df = pd.DataFrame(feature_stats)
        results_df = results_df.sort_values('composite_score', ascending=False)
        
        n_select = min(10, len(results_df))
        selected_features = clinical_features + results_df.head(n_select)['Gene'].tolist()
        
        return selected_features
    
    def train_evaluate_iteration(self, X_train, X_test, y_train, y_test, selected_features):
        try:
            X_train_selected = X_train[selected_features]
            X_test_selected = X_test[selected_features]
            
            imputer = SimpleImputer(strategy='median')
            scaler = StandardScaler()
            
            categorical_cols = [col for col in selected_features if X_train_selected[col].dtype == 'object']
            
            if categorical_cols:
                for col in categorical_cols:
                    X_train_selected[col] = pd.Categorical(X_train_selected[col]).codes
                    X_test_selected[col] = pd.Categorical(X_test_selected[col]).codes
            
            X_train_processed = scaler.fit_transform(imputer.fit_transform(X_train_selected))
            X_test_processed = scaler.transform(imputer.transform(X_test_selected))
            
            model = LogisticRegression(C=1.0, penalty='l1', solver='liblinear', 
                                     class_weight='balanced', random_state=42)
            model.fit(X_train_processed, y_train)
            
            y_pred = model.predict(X_test_processed)
            y_pred_proba = model.predict_proba(X_test_processed)[:, 1]
            
            auc = roc_auc_score(y_test, y_pred_proba)
            accuracy = accuracy_score(y_test, y_pred)
            
            return {'auc': auc, 'accuracy': accuracy, 'n_features': len(selected_features)}
            
        except Exception:
            return None
    
    def create_preprocessing_pipeline(self, stable_features):
        print("\n" + "=" * 50)
        print("STEP 5: CREATE PREPROCESSING PIPELINE")
        print("=" * 50)
        
        clinical_features = [f for f in stable_features if f not in self.methylation_genes]
        methylation_features = [f for f in stable_features if f in self.methylation_genes]
        
        numerical_features = []
        categorical_features = []
        
        for feature in clinical_features:
            if feature in ['ethnicity', 'v3n_Gender', 'BMI_cat']:
                categorical_features.append(feature)
            else:
                numerical_features.append(feature)
        
        print(f"Numerical features: {numerical_features}")
        print(f"Categorical features: {categorical_features}")
        print(f"Selected methylation features: {len(methylation_features)}")
        
        preprocessors = []
        
        if len(numerical_features + methylation_features) > 0:
            numerical_pipeline = Pipeline([
                ('imputer', SimpleImputer(strategy='median')),
                ('scaler', StandardScaler())
            ])
            preprocessors.append(('numerical', numerical_pipeline, 
                                numerical_features + methylation_features))
        
        if len(categorical_features) > 0:
            categorical_pipeline = Pipeline([
                ('imputer', SimpleImputer(strategy='most_frequent')),
                ('onehot', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'))
            ])
            preprocessors.append(('categorical', categorical_pipeline, categorical_features))
        
        if len(preprocessors) > 1:
            self.preprocessor = ColumnTransformer(preprocessors, remainder='drop')
        else:
            self.preprocessor = Pipeline([
                ('imputer', SimpleImputer(strategy='median')),
                ('scaler', StandardScaler())
            ])
        
        print(f"Created preprocessing pipeline with {len(preprocessors)} transformers")
        return True
    
    def train_final_model(self):
        print("\n" + "=" * 50)
        print("STEP 6: TRAIN FINAL MODEL")
        print("=" * 50)
        
        param_grid = {
            'model__C': [0.001, 0.01, 0.1, 1.0, 10.0],
            'model__penalty': ['l1'],
            'model__solver': ['liblinear'],
            'model__class_weight': ['balanced']
        }
        
        pipeline = Pipeline([
            ('preprocessor', self.preprocessor),
            ('model', LogisticRegression(random_state=42))
        ])
        
        cv_inner = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
        
        grid_search = GridSearchCV(
            pipeline, param_grid, cv=cv_inner,
            scoring='roc_auc', n_jobs=1, verbose=1
        )
        
        print("Starting hyperparameter optimization...")
        grid_search.fit(self.X_train, self.y_train)
        
        print(f"Best CV score: {grid_search.best_score_:.4f}")
        print(f"Best parameters: {grid_search.best_params_}")
        
        self.model = grid_search.best_estimator_
        return True
    
    def evaluate_final_model(self):
        print("\n" + "=" * 50)
        print("STEP 7: EVALUATE FINAL MODEL")
        print("=" * 50)
        
        y_pred = self.model.predict(self.X_test)
        y_pred_proba = self.model.predict_proba(self.X_test)[:, 1]
        
        auc_score = roc_auc_score(self.y_test, y_pred_proba)
        accuracy = accuracy_score(self.y_test, y_pred)
        f1 = f1_score(self.y_test, y_pred)
        
        print(f"Test AUC: {auc_score:.4f}")
        print(f"Test Accuracy: {accuracy:.4f}")
        print(f"Test F1-Score: {f1:.4f}")
        
        cm = confusion_matrix(self.y_test, y_pred)
        print(f"\nConfusion Matrix:")
        print(cm)
        
        if cm.size == 4:
            tn, fp, fn, tp = cm.ravel()
            precision = tp / (tp + fp) if (tp + fp) > 0 else 0
            recall = tp / (tp + fn) if (tp + fn) > 0 else 0
            specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
            
            print(f"\nPrecision: {precision:.4f}")
            print(f"Recall: {recall:.4f}")
            print(f"Specificity: {specificity:.4f}")
        else:
            precision = recall = specificity = 0
        
        cv_scores = cross_val_score(self.model, self.X_train, self.y_train,
                                   cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
                                   scoring='roc_auc')
        
        print(f"\nCV scores: {[f'{score:.3f}' for score in cv_scores]}")
        print(f"Mean CV AUC: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
        
        evaluation_results = {
            'test_metrics': {
                'auc_roc': auc_score,
                'accuracy': accuracy,
                'f1_score': f1,
                'precision': precision,
                'recall_sensitivity': recall,
                'specificity': specificity
            },
            'confusion_matrix': cm.tolist(),
            'cross_validation': {
                'cv_scores': cv_scores.tolist(),
                'mean_cv_auc': cv_scores.mean(),
                'std_cv_auc': cv_scores.std()
            },
            'predictions': {
                'y_test': self.y_test.tolist(),
                'y_pred': y_pred.tolist(),
                'y_pred_proba': y_pred_proba.tolist()
            }
        }
        
        return evaluation_results
    
    def save_results(self, multiple_splits_results, evaluation_results):
        print("\n" + "=" * 50)
        print("STEP 8: SAVE RESULTS")
        print("=" * 50)
        
        model_file = os.path.join(self.output_dir, f"{self.prefix}trained_model_{self.timestamp}.joblib")
        joblib.dump(self.model, model_file)
        print(f"Model saved to: {model_file}")
        
        encoder_file = os.path.join(self.output_dir, f"{self.prefix}label_encoder_{self.timestamp}.joblib")
        joblib.dump(self.label_encoder, encoder_file)
        
        evaluation_file = os.path.join(self.output_dir, f"{self.prefix}evaluation_results_{self.timestamp}.json")
        combined_results = {
            'evaluation_metrics': evaluation_results,
            'multiple_splits_analysis': multiple_splits_results,
            'selected_features': self.feature_names,
            'timestamp': self.timestamp
        }
        
        with open(evaluation_file, 'w') as f:
            json.dump(combined_results, f, indent=4)
        print(f"Results saved to: {evaluation_file}")
        
        return True
    
    def run_complete_pipeline(self, clinical_path, dmr_path, methylation_path):
        print("\n" + "=" * 80)
        print("STARTING B12 PREDICTION - MULTIPLE SPLITS APPROACH")
        print("=" * 80)
        
        pipeline_start_time = datetime.now()
        
        try:
            if not self.load_data(clinical_path, dmr_path, methylation_path):
                print("ERROR: Data loading failed")
                return False
            
            if not self.prepare_combined_dataset():
                print("ERROR: Dataset preparation failed")
                return False
            
            multiple_splits_results = self.run_multiple_train_test_splits(n_iterations=20)
            
            multiple_splits_results = self.run_multiple_train_test_splits(n_iterations=20)
            
            stable_features = multiple_splits_results['stable_features_ranking'][:15]
            self.selected_methylation_genes = [feature for feature, _ in stable_features 
                                             if feature in self.methylation_genes]
            
            clinical_features = [f for f in self.feature_names if f not in self.methylation_genes]
            self.feature_names = clinical_features + self.selected_methylation_genes
            
            print(f"\nUsing {len(self.selected_methylation_genes)} most stable methylation features for final model")
            
            self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
                self.X[self.feature_names], self.y, test_size=0.2, random_state=42, stratify=self.y
            )
            
            if not self.create_preprocessing_pipeline(self.feature_names):
                print("ERROR: Preprocessing pipeline creation failed")
                return False
            
            if not self.train_final_model():
                print("ERROR: Model training failed")
                return False
            
            evaluation_results = self.evaluate_final_model()
            
            if not self.save_results(multiple_splits_results, evaluation_results):
                print("ERROR: Saving results failed")
                return False
            
            pipeline_end_time = datetime.now()
            pipeline_duration = pipeline_end_time - pipeline_start_time
            
            print("\n" + "=" * 80)
            print("PIPELINE COMPLETION SUMMARY")
            print("=" * 80)
            print(f"Total duration: {pipeline_duration}")
            
            print("\nMULTIPLE SPLITS ANALYSIS RESULTS:")
            splits_results = multiple_splits_results
            print(f"Total iterations: {splits_results['n_iterations']}")
            print(f"Successful iterations: {splits_results['successful_iterations']}")
            
            if splits_results['performance_stats']['mean_auc']:
                mean_auc = splits_results['performance_stats']['mean_auc']
                std_auc = splits_results['performance_stats']['std_auc']
                print(f"Mean AUC across iterations: {mean_auc:.3f} ± {std_auc:.3f}")
                
                if std_auc < 0.1:
                    print("Low variance - Model performance is consistent")
                elif std_auc < 0.2:
                    print("Moderate variance - Some performance variability")
                else:
                    print("High variance - Performance highly dependent on data split")
            
            stable_features_top10 = splits_results['stable_features_ranking'][:10]
            highly_stable = splits_results['stability_insights']['highly_stable_features']
            
            print(f"\nTop 10 most stable features:")
            for i, (feature, count) in enumerate(stable_features_top10, 1):
                rate = (count / splits_results['successful_iterations']) * 100
                stability_marker = "HIGHLY STABLE" if rate >= 70 else "STABLE" if rate >= 50 else "VARIABLE"
                print(f"  {i:2d}. {feature:<15} ({rate:3.0f}% selection) - {stability_marker}")
            
            print(f"\nFeature Selection Stability:")
            if highly_stable >= 5:
                print(f"ROBUST: {highly_stable} features consistently selected (>=70% iterations)")
            elif highly_stable >= 2:
                print(f"MODERATE: {highly_stable} features consistently selected")
            else:
                print(f"UNSTABLE: Only {highly_stable} highly stable features")
            
            print(f"\nFinal Model Performance:")
            auc = evaluation_results['test_metrics']['auc_roc']
            accuracy = evaluation_results['test_metrics']['accuracy']
            print(f"Hold-out Test AUC: {auc:.4f}")
            print(f"Hold-out Test Accuracy: {accuracy:.4f}")
            
            cv_auc = evaluation_results['cross_validation']['mean_cv_auc']
            cv_std = evaluation_results['cross_validation']['std_cv_auc']
            print(f"Cross-validation AUC: {cv_auc:.4f} ± {cv_std:.4f}")
            
            print(f"\nRecommended Focus: The {highly_stable} most stable features for future validation")
            print("=" * 80)
            
            return True
            
        except Exception as e:
            print(f"ERROR: Pipeline failed with exception: {str(e)}")
            self.logger.error(f"Pipeline failed with exception: {str(e)}")
            import traceback
            traceback.print_exc()
            return False


def main():
    """Main function to run the data-driven B12 prediction model"""
    
    clinical_path = "/Users/heweilin/Desktop/P056/7Clinical_data50.csv"
    dmr_path = "/Users/heweilin/Desktop/P056/4DNA_DMRs.csv"
    methylation_path = "/Users/heweilin/Desktop/P056_Code_2/Processed_Data/1_PD_PromoterRegion_CpGs.csv"
    output_dir = "/Users/heweilin/Desktop/P056_Code_3/Data"
    
    print("B12 Deficiency Prediction Model - Multiple Splits Approach")
    print("=" * 70)
    print("Key Features:")
    print("- Multiple independent train-test splits (20 iterations)")
    print("- Batch correction for methylation data")
    print("- Feature stability analysis across iterations")
    print("- Mann-Whitney U + Cohen's d feature selection")
    print("- LASSO logistic regression with L1 regularization")
    print("- Robust identification of stable biomarkers")
    print("=" * 70)
    print(f"Clinical data: {clinical_path}")
    print(f"Methylation data: {methylation_path}")
    print(f"Output directory: {output_dir}")
    print("=" * 70)
    
    model = B12PredictionDataDriven(output_dir=output_dir, prefix="MultiSplits_")
    
    success = model.run_complete_pipeline(clinical_path, dmr_path, methylation_path)
    
    if success:
        print("\n" + "=" * 80)
        print("MULTIPLE SPLITS MODEL PIPELINE COMPLETED SUCCESSFULLY!")
        print("=" * 80)
        print("Key Achievements:")
        print("- 20 independent train-test splits completed")
        print("- Feature stability analysis identifies robust biomarkers")
        print("- Batch correction applied to methylation data")
        print("- No data leakage - each split uses independent feature selection")
        print("- Final model trained on most stable features")
        print("- Comprehensive evaluation with realistic performance estimates")
        print("\nCheck the output directory for detailed results and analysis.")
        print("=" * 80)
    else:
        print("\nMULTIPLE SPLITS MODEL PIPELINE FAILED!")
        print("Check the log files for detailed error information.")
    
    return success


if __name__ == "__main__":
    np.random.seed(42)
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
B12 Deficiency Prediction Model - Mechanism-Based Feature Selection
Using predefined CpG-mRNA negative regulatory pairs

Author: Research Team
Date: August 25, 2025
Version: 1.0 - Mechanism-Based Approach with Biologically Informed Features

Strategy: Use known negative regulatory CpG-mRNA pairs as features
Objective: Predict B12 deficiency using biologically meaningful methylation features
"""

import os
import sys
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import (train_test_split, StratifiedKFold, 
                                   cross_val_score, GridSearchCV)
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (classification_report, confusion_matrix, 
                           roc_auc_score, roc_curve, precision_recall_curve,
                           f1_score, accuracy_score)
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json
from datetime import datetime
import logging

class B12PredictionMechanismBased:
    """B12 deficiency prediction using mechanism-based features (CpG-mRNA regulatory pairs)"""
    
    def __init__(self, output_dir="/Users/heweilin/Desktop/P056_Code_3/Data", prefix="Mechanism_"):
        self.output_dir = output_dir
        self.prefix = prefix
        self.timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        
        os.makedirs(output_dir, exist_ok=True)
        self.setup_logging()
        
        # Initialize data storage
        self.clinical_data = None
        self.methylation_data = None
        self.dmr_data = None
        self.X = None
        self.y = None
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        self.feature_names = None
        self.model = None
        self.preprocessor = None
        self.label_encoder = None
        
        # Define the 10 mechanism-based CpG-mRNA regulatory pairs
        self.cpg_mrna_pairs = {
            'IQCG': {'beta': -0.889, 'fdr_p': 6.11e-11, 'r_squared': 0.785, 'effect_size': 'Large'},
            'ZNF154': {'beta': -0.923, 'fdr_p': 3.97e-8, 'r_squared': 0.727, 'effect_size': 'Large'},
            'TSTD1': {'beta': -0.771, 'fdr_p': 3.97e-8, 'r_squared': 0.702, 'effect_size': 'Large'},
            'GTSF1': {'beta': -0.804, 'fdr_p': 3.97e-8, 'r_squared': 0.735, 'effect_size': 'Large'},
            'SEPTIN7P11': {'beta': -0.852, 'fdr_p': 7.70e-8, 'r_squared': 0.704, 'effect_size': 'Large'},
            'NUDT10': {'beta': -0.920, 'fdr_p': 2.82e-7, 'r_squared': 0.717, 'effect_size': 'Large'},
            'NAA11': {'beta': -0.722, 'fdr_p': 6.75e-7, 'r_squared': 0.684, 'effect_size': 'Large'},
            'SYCP3': {'beta': -0.692, 'fdr_p': 2.35e-6, 'r_squared': 0.751, 'effect_size': 'Large'},
            'PSMA8': {'beta': -0.716, 'fdr_p': 2.26e-5, 'r_squared': 0.591, 'effect_size': 'Large'},
            'LINC00667': {'beta': -0.797, 'fdr_p': 2.26e-5, 'r_squared': 0.550, 'effect_size': 'Large'}
        }
        
        print("=" * 80)
        print("B12 DEFICIENCY PREDICTION - MECHANISM-BASED APPROACH")
        print("=" * 80)
        print(f"Timestamp: {self.timestamp}")
        print(f"Output directory: {output_dir}")
        print("Key Features:")
        print("- 10 predefined CpG-mRNA negative regulatory pairs")
        print("- Biologically informed feature selection")
        print("- Clinical covariates + mechanism-based methylation features")
        print("- Multiple train-test splits for validation")
        print("- Batch correction for methylation data")
        print("=" * 80)
        
        print("\nPredefined CpG-mRNA Regulatory Pairs:")
        print("Rank | Gene Symbol | Beta    | FDR p-value | R²    | Effect")
        print("-" * 60)
        for i, (gene, stats) in enumerate(self.cpg_mrna_pairs.items(), 1):
            print(f"{i:2d}   | {gene:<11} | {stats['beta']:>6.3f} | {stats['fdr_p']:>11.2e} | {stats['r_squared']:>5.3f} | {stats['effect_size']}")
        print("=" * 80)
    
    def setup_logging(self):
        log_file = os.path.join(self.output_dir, f"{self.prefix}model_log_{self.timestamp}.txt")
        logging.basicConfig(
            level=logging.INFO,
            format='%(asctime)s - %(levelname)s - %(message)s',
            handlers=[
                logging.FileHandler(log_file),
                logging.StreamHandler(sys.stdout)
            ]
        )
        self.logger = logging.getLogger(__name__)
        self.logger.info("Mechanism-based model logging setup completed")
    
    def load_data(self, clinical_path, dmr_path, methylation_path):
        print("\n" + "=" * 50)
        print("STEP 1: DATA LOADING")
        print("=" * 50)
        
        print(f"Loading clinical data from: {clinical_path}")
        try:
            self.clinical_data = pd.read_csv(clinical_path)
            print(f"Clinical data shape: {self.clinical_data.shape}")
            self.logger.info(f"Clinical data loaded: {self.clinical_data.shape}")
        except Exception as e:
            print(f"Error loading clinical data: {e}")
            return False
        
        print(f"Loading DMR data from: {dmr_path}")
        try:
            self.dmr_data = pd.read_csv(dmr_path, low_memory=False)
            print(f"DMR data shape: {self.dmr_data.shape}")
            self.logger.info(f"DMR data loaded: {self.dmr_data.shape}")
        except Exception as e:
            print(f"Error loading DMR data: {e}")
            return False
        
        print(f"Loading methylation data from: {methylation_path}")
        try:
            self.methylation_data = pd.read_csv(methylation_path)
            print(f"Methylation data shape: {self.methylation_data.shape}")
            self.logger.info(f"Methylation data loaded: {self.methylation_data.shape}")
        except Exception as e:
            print(f"Error loading methylation data: {e}")
            return False
        
        return True
    
    def prepare_methylation_data_with_batch_correction(self):
        print("\n" + "=" * 50)
        print("STEP 2: BATCH CORRECTION")
        print("=" * 50)
        
        methylation_df = self.methylation_data.set_index('Gene_Symbol').T
        print(f"Methylation data after transpose: {methylation_df.shape}")
        
        clinical_ids = self.clinical_data['DNA_ID'].values
        methylation_samples = methylation_df.index.values
        
        common_samples = list(set(clinical_ids) & set(methylation_samples))
        print(f"Common samples: {len(common_samples)}")
        
        if len(common_samples) == 0:
            print("Warning: No common samples found")
            return None, None
        
        methylation_filtered = methylation_df.loc[common_samples]
        clinical_aligned = self.clinical_data[self.clinical_data['DNA_ID'].isin(common_samples)]
        clinical_aligned = clinical_aligned.set_index('DNA_ID').loc[common_samples]
        
        if 'batch_DNA' in clinical_aligned.columns:
            batch_info = clinical_aligned['batch_DNA'].values
            unique_batches = np.unique(batch_info[~pd.isna(batch_info)])
            print(f"Found DNA batches: {unique_batches}")
            print(f"Batch distribution: {pd.Series(batch_info).value_counts().to_dict()}")
            
            print("Performing batch correction...")
            methylation_corrected = self.apply_batch_correction(methylation_filtered, batch_info)
            print(f"Batch-corrected data shape: {methylation_corrected.shape}")
        else:
            print("No batch information found, skipping correction")
            methylation_corrected = methylation_filtered
        
        return methylation_corrected, common_samples
    
    def apply_batch_correction(self, methylation_data, batch_info):
        corrected_data = methylation_data.copy()
        batches = pd.Categorical(batch_info)
        unique_batches = batches.categories
        
        if len(unique_batches) <= 1:
            print("Only one batch found, no correction needed")
            return corrected_data
        
        print(f"Correcting for {len(unique_batches)} batches")
        n_genes_corrected = 0
        
        for gene in corrected_data.columns:
            gene_values = corrected_data[gene].values
            
            if np.sum(~np.isnan(gene_values)) < len(gene_values) * 0.5:
                continue
            if np.nanvar(gene_values) == 0:
                continue
            
            try:
                corrected_values = self.correct_gene_for_batch(gene_values, batches)
                corrected_data.loc[:, gene] = corrected_values
                n_genes_corrected += 1
            except Exception:
                continue
        
        print(f"Successfully corrected {n_genes_corrected} genes for batch effects")
        return corrected_data
    
    def correct_gene_for_batch(self, gene_values, batches):
        df = pd.DataFrame({'values': gene_values, 'batch': batches})
        df_clean = df.dropna()
        
        if len(df_clean) < 3:
            return gene_values
        
        overall_mean = df_clean['values'].mean()
        batch_means = df_clean.groupby('batch')['values'].mean()
        batch_effects = batch_means - overall_mean
        
        corrected_values = gene_values.copy()
        for batch_name, batch_effect in batch_effects.items():
            batch_mask = (batches == batch_name)
            corrected_values[batch_mask] = corrected_values[batch_mask] - batch_effect
        
        return corrected_values
    
    def extract_mechanism_based_features(self):
        print("\n" + "=" * 50)
        print("STEP 3: EXTRACT MECHANISM-BASED METHYLATION FEATURES")
        print("=" * 50)
        
        methylation_result = self.prepare_methylation_data_with_batch_correction()
        if methylation_result[0] is None:
            print("Error: Could not prepare methylation data")
            return None, None
        
        methylation_X, common_samples = methylation_result
        
        # Check which of our target genes are available in the methylation data
        target_genes = list(self.cpg_mrna_pairs.keys())
        available_genes = []
        missing_genes = []
        
        print("Checking availability of mechanism-based features:")
        for gene in target_genes:
            if gene in methylation_X.columns:
                available_genes.append(gene)
                stats = self.cpg_mrna_pairs[gene]
                print(f"  ✓ {gene:<12} (β={stats['beta']:>6.3f}, R²={stats['r_squared']:>5.3f})")
            else:
                missing_genes.append(gene)
                print(f"  ✗ {gene:<12} (NOT FOUND in methylation data)")
        
        print(f"\nSummary:")
        print(f"  Available mechanism-based features: {len(available_genes)}")
        print(f"  Missing features: {len(missing_genes)}")
        
        if len(available_genes) == 0:
            print("ERROR: No mechanism-based features found in methylation data")
            return None, None
        
        # Extract only the available mechanism-based methylation features
        mechanism_methylation_features = methylation_X[available_genes]
        
        print(f"Extracted mechanism-based methylation features shape: {mechanism_methylation_features.shape}")
        
        return mechanism_methylation_features, common_samples, available_genes, missing_genes
    
    def prepare_combined_dataset(self):
        print("\n" + "=" * 50)
        print("STEP 4: COMBINE CLINICAL AND MECHANISM-BASED FEATURES")
        print("=" * 50)
        
        # Clinical features (excluding B12_mol and batch_DNA)
        numerical_features = ['age', 'BMI', 'parity', 'B12supplem', 
                             'v1p_MultivitTab', 'v1p_FolicAcid', 'smoking']
        categorical_features = ['ethnicity', 'v3n_Gender', 'BMI_cat']
        
        available_numerical = [f for f in numerical_features if f in self.clinical_data.columns]
        available_categorical = [f for f in categorical_features if f in self.clinical_data.columns]
        
        print(f"Available clinical features:")
        print(f"  Numerical: {available_numerical}")
        print(f"  Categorical: {available_categorical}")
        print("EXCLUDED: B12_mol (target leakage), batch_DNA (used for correction)")
        
        all_clinical_features = available_numerical + available_categorical
        
        # Extract mechanism-based methylation features
        methylation_result = self.extract_mechanism_based_features()
        if methylation_result[0] is None:
            print("ERROR: Could not extract mechanism-based features")
            return False
        
        mechanism_methylation_X, common_samples, available_genes, missing_genes = methylation_result
        
        # Align clinical data with common samples
        clinical_filtered = self.clinical_data[self.clinical_data['DNA_ID'].isin(common_samples)]
        clinical_filtered = clinical_filtered.set_index('DNA_ID').loc[common_samples]
        clinical_X_filtered = clinical_filtered[all_clinical_features]
        
        # Combine clinical and mechanism-based methylation features
        self.X = pd.concat([clinical_X_filtered, mechanism_methylation_X], axis=1)
        self.feature_names = list(clinical_X_filtered.columns) + available_genes
        
        # Store target variable
        self.y = clinical_filtered['B12_status'].values
        
        print(f"Final combined dataset:")
        print(f"  Total samples: {self.X.shape[0]}")
        print(f"  Clinical features: {len(clinical_X_filtered.columns)}")
        print(f"  Mechanism-based methylation features: {len(available_genes)}")
        print(f"  Total features: {len(self.feature_names)}")
        
        # Store information about available and missing genes
        self.available_mechanism_genes = available_genes
        self.missing_mechanism_genes = missing_genes
        
        # Encode target variable
        self.label_encoder = LabelEncoder()
        self.y = self.label_encoder.fit_transform(self.y)
        print(f"Target encoding: {dict(zip(self.label_encoder.classes_, self.label_encoder.transform(self.label_encoder.classes_)))}")
        
        return True
    
    def run_multiple_train_test_splits(self, n_iterations=20, test_size=0.2):
        print("\n" + "=" * 50)
        print("STEP 5: MULTIPLE TRAIN-TEST SPLITS WITH MECHANISM-BASED FEATURES")
        print("=" * 50)
        
        print(f"Running {n_iterations} independent train-test splits")
        print(f"Using predefined mechanism-based features (no feature selection)")
        print(f"Each split: {(1-test_size)*100:.0f}% training, {test_size*100:.0f}% testing")
        
        all_performances = []
        iteration_details = []
        feature_importance_across_iterations = {gene: [] for gene in self.available_mechanism_genes}
        
        for iteration in range(n_iterations):
            print(f"\nIteration {iteration + 1}/{n_iterations}")
            
            X_iter_train, X_iter_test, y_iter_train, y_iter_test = train_test_split(
                self.X, self.y, test_size=test_size,
                random_state=42 + iteration,
                stratify=self.y
            )
            
            print(f"Train: {len(X_iter_train)} samples, Test: {len(X_iter_test)} samples")
            
            try:
                performance, feature_importance = self.train_evaluate_iteration(
                    X_iter_train, X_iter_test, y_iter_train, y_iter_test
                )
                
                all_performances.append(performance)
                
                # Store feature importance for mechanism-based features
                for gene in self.available_mechanism_genes:
                    if gene in feature_importance:
                        feature_importance_across_iterations[gene].append(feature_importance[gene])
                
                iteration_details.append({
                    'iteration': iteration + 1,
                    'performance': performance,
                    'feature_importance': feature_importance,
                    'n_train': len(X_iter_train),
                    'n_test': len(X_iter_test)
                })
                
                print(f"AUC: {performance['auc']:.3f}, Accuracy: {performance['accuracy']:.3f}")
                
            except Exception as e:
                print(f"Error in iteration {iteration + 1}: {e}")
                continue
        
        # Analyze performance consistency
        print(f"\nPERFORMANCE CONSISTENCY ANALYSIS ({len(all_performances)} successful iterations)")
        print("=" * 60)
        
        valid_performances = [p for p in all_performances if p is not None]
        if valid_performances:
            aucs = [p['auc'] for p in valid_performances]
            accuracies = [p['accuracy'] for p in valid_performances]
            
            print(f"Performance Statistics:")
            print(f"  Mean AUC: {np.mean(aucs):.3f} ± {np.std(aucs):.3f}")
            print(f"  AUC Range: [{np.min(aucs):.3f}, {np.max(aucs):.3f}]")
            print(f"  Mean Accuracy: {np.mean(accuracies):.3f} ± {np.std(accuracies):.3f}")
            print(f"  Iterations with AUC > 0.8: {sum(1 for auc in aucs if auc > 0.8)}/{len(aucs)}")
            print(f"  Iterations with AUC > 0.9: {sum(1 for auc in aucs if auc > 0.9)}/{len(aucs)}")
            
            # Performance consistency assessment
            auc_std = np.std(aucs)
            if auc_std < 0.1:
                consistency_level = "HIGH - Model performance is consistent"
            elif auc_std < 0.2:
                consistency_level = "MODERATE - Some performance variability"
            else:
                consistency_level = "LOW - High performance variability"
            
            print(f"  Performance Consistency: {consistency_level}")
        
        # Analyze mechanism-based feature importance consistency
        print(f"\nMECHANISM-BASED FEATURE IMPORTANCE CONSISTENCY:")
        print("=" * 60)
        
        feature_importance_summary = {}
        for gene in self.available_mechanism_genes:
            importances = feature_importance_across_iterations[gene]
            if importances:
                feature_importance_summary[gene] = {
                    'mean_importance': np.mean(importances),
                    'std_importance': np.std(importances),
                    'consistency_score': 1 - (np.std(importances) / (np.mean(np.abs(importances)) + 1e-8))
                }
        
        # Sort by mean importance
        sorted_features = sorted(feature_importance_summary.items(), 
                               key=lambda x: abs(x[1]['mean_importance']), reverse=True)
        
        print("Feature Importance Ranking (across all iterations):")
        print("Rank | Gene Symbol | Mean Coef | Std Dev  | Consistency | Original Stats")
        print("-" * 75)
        for i, (gene, stats) in enumerate(sorted_features, 1):
            orig_stats = self.cpg_mrna_pairs[gene]
            consistency_icon = "★" if stats['consistency_score'] > 0.7 else "◆" if stats['consistency_score'] > 0.5 else "○"
            print(f"{i:2d}   | {gene:<11} | {stats['mean_importance']:>8.3f} | {stats['std_importance']:>8.3f} | "
                  f"{consistency_icon} {stats['consistency_score']:>5.3f}  | β={orig_stats['beta']:>6.3f}, R²={orig_stats['r_squared']:>5.3f}")
        
        results_summary = {
            'n_iterations': n_iterations,
            'successful_iterations': len(all_performances),
            'available_mechanism_genes': self.available_mechanism_genes,
            'missing_mechanism_genes': self.missing_mechanism_genes,
            'performance_stats': {
                'mean_auc': np.mean(aucs) if valid_performances else None,
                'std_auc': np.std(aucs) if valid_performances else None,
                'mean_accuracy': np.mean(accuracies) if valid_performances else None,
                'auc_range': [np.min(aucs), np.max(aucs)] if valid_performances else None,
                'consistency_level': consistency_level if valid_performances else None
            },
            'feature_importance_summary': feature_importance_summary,
            'iteration_details': iteration_details
        }
        
        results_file = os.path.join(self.output_dir, f"{self.prefix}multiple_splits_analysis_{self.timestamp}.json")
        with open(results_file, 'w') as f:
            json.dump(results_summary, f, indent=4)
        print(f"\nComplete analysis saved to: {results_file}")
        
        return results_summary
    
    def train_evaluate_iteration(self, X_train, X_test, y_train, y_test):
        try:
            # Simple preprocessing
            imputer = SimpleImputer(strategy='median')
            scaler = StandardScaler()
            
            # Handle categorical features
            categorical_cols = [col for col in X_train.columns if X_train[col].dtype == 'object']
            
            X_train_processed = X_train.copy()
            X_test_processed = X_test.copy()
            
            if categorical_cols:
                for col in categorical_cols:
                    X_train_processed[col] = pd.Categorical(X_train_processed[col]).codes
                    X_test_processed[col] = pd.Categorical(X_test_processed[col]).codes
            
            # Apply preprocessing
            X_train_final = scaler.fit_transform(imputer.fit_transform(X_train_processed))
            X_test_final = scaler.transform(imputer.transform(X_test_processed))
            
            # Train model
            model = LogisticRegression(C=1.0, penalty='l1', solver='liblinear',
                                     class_weight='balanced', random_state=42)
            model.fit(X_train_final, y_train)
            
            # Evaluate
            y_pred = model.predict(X_test_final)
            y_pred_proba = model.predict_proba(X_test_final)[:, 1]
            
            auc = roc_auc_score(y_test, y_pred_proba)
            accuracy = accuracy_score(y_test, y_pred)
            
            # Extract feature importance (coefficients)
            feature_importance = {}
            for i, feature_name in enumerate(X_train.columns):
                feature_importance[feature_name] = model.coef_[0][i]
            
            return {
                'auc': auc,
                'accuracy': accuracy,
                'n_features': len(X_train.columns)
            }, feature_importance
            
        except Exception as e:
            return None, {}
    
    def create_preprocessing_pipeline(self):
        print("\n" + "=" * 50)
        print("STEP 6: CREATE PREPROCESSING PIPELINE")
        print("=" * 50)
        
        clinical_features = [f for f in self.feature_names if f not in self.available_mechanism_genes]
        
        numerical_features = []
        categorical_features = []
        
        for feature in clinical_features:
            if feature in ['ethnicity', 'v3n_Gender', 'BMI_cat']:
                categorical_features.append(feature)
            else:
                numerical_features.append(feature)
        
        print(f"Numerical features: {numerical_features}")
        print(f"Categorical features: {categorical_features}")
        print(f"Mechanism-based methylation features: {len(self.available_mechanism_genes)}")
        
        preprocessors = []
        
        if len(numerical_features + self.available_mechanism_genes) > 0:
            numerical_pipeline = Pipeline([
                ('imputer', SimpleImputer(strategy='median')),
                ('scaler', StandardScaler())
            ])
            preprocessors.append(('numerical', numerical_pipeline,
                                numerical_features + self.available_mechanism_genes))
        
        if len(categorical_features) > 0:
            categorical_pipeline = Pipeline([
                ('imputer', SimpleImputer(strategy='most_frequent')),
                ('onehot', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'))
            ])
            preprocessors.append(('categorical', categorical_pipeline, categorical_features))
        
        if len(preprocessors) > 1:
            self.preprocessor = ColumnTransformer(preprocessors, remainder='drop')
        else:
            self.preprocessor = Pipeline([
                ('imputer', SimpleImputer(strategy='median')),
                ('scaler', StandardScaler())
            ])
        
        print(f"Created preprocessing pipeline with {len(preprocessors)} transformers")
        return True
    
    def train_final_model(self):
        print("\n" + "=" * 50)
        print("STEP 7: TRAIN FINAL MODEL")
        print("=" * 50)
        
        param_grid = {
            'model__C': [0.001, 0.01, 0.1, 1.0, 10.0],
            'model__penalty': ['l1'],
            'model__solver': ['liblinear'],
            'model__class_weight': ['balanced']
        }
        
        pipeline = Pipeline([
            ('preprocessor', self.preprocessor),
            ('model', LogisticRegression(random_state=42))
        ])
        
        cv_inner = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
        
        grid_search = GridSearchCV(
            pipeline, param_grid, cv=cv_inner,
            scoring='roc_auc', n_jobs=1, verbose=1
        )
        
        print("Starting hyperparameter optimization...")
        grid_search.fit(self.X_train, self.y_train)
        
        print(f"Best CV score: {grid_search.best_score_:.4f}")
        print(f"Best parameters: {grid_search.best_params_}")
        
        self.model = grid_search.best_estimator_
        return True
    
    def evaluate_final_model(self):
        print("\n" + "=" * 50)
        print("STEP 8: EVALUATE FINAL MODEL")
        print("=" * 50)
        
        y_pred = self.model.predict(self.X_test)
        y_pred_proba = self.model.predict_proba(self.X_test)[:, 1]
        
        auc_score = roc_auc_score(self.y_test, y_pred_proba)
        accuracy = accuracy_score(self.y_test, y_pred)
        f1 = f1_score(self.y_test, y_pred)
        
        print(f"Final Model Performance:")
        print(f"  Test AUC: {auc_score:.4f}")
        print(f"  Test Accuracy: {accuracy:.4f}")
        print(f"  Test F1-Score: {f1:.4f}")
        
        cm = confusion_matrix(self.y_test, y_pred)
        print(f"\nConfusion Matrix:")
        print(cm)
        
        if cm.size == 4:
            tn, fp, fn, tp = cm.ravel()
            precision = tp / (tp + fp) if (tp + fp) > 0 else 0
            recall = tp / (tp + fn) if (tp + fn) > 0 else 0
            specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
            
            print(f"\nDetailed Metrics:")
            print(f"  Precision: {precision:.4f}")
            print(f"  Recall: {recall:.4f}")
            print(f"  Specificity: {specificity:.4f}")
        else:
            precision = recall = specificity = 0
        
        cv_scores = cross_val_score(self.model, self.X_train, self.y_train,
                                   cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
                                   scoring='roc_auc')
        
        print(f"\nCross-validation Performance:")
        print(f"  CV scores: {[f'{score:.3f}' for score in cv_scores]}")
        print(f"  Mean CV AUC: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
        
        evaluation_results = {
            'test_metrics': {
                'auc_roc': auc_score,
                'accuracy': accuracy,
                'f1_score': f1,
                'precision': precision,
                'recall_sensitivity': recall,
                'specificity': specificity
            },
            'confusion_matrix': cm.tolist(),
            'cross_validation': {
                'cv_scores': cv_scores.tolist(),
                'mean_cv_auc': cv_scores.mean(),
                'std_cv_auc': cv_scores.std()
            },
            'predictions': {
                'y_test': self.y_test.tolist(),
                'y_pred': y_pred.tolist(),
                'y_pred_proba': y_pred_proba.tolist()
            }
        }
        
        return evaluation_results
    
    def analyze_mechanism_feature_importance(self):
        print("\n" + "=" * 50)
        print("STEP 9: MECHANISM-BASED FEATURE IMPORTANCE ANALYSIS")
        print("=" * 50)
        
        try:
            lr_model = self.model.named_steps['model']
            coefficients = lr_model.coef_[0]
            
            # Get feature names after preprocessing
            try:
                if hasattr(self.model.named_steps['preprocessor'], 'get_feature_names_out'):
                    transformed_feature_names = self.model.named_steps['preprocessor'].get_feature_names_out()
                else:
                    transformed_feature_names = [f"feature_{i}" for i in range(len(coefficients))]
            except:
                transformed_feature_names = [f"feature_{i}" for i in range(len(coefficients))]
            
            # Create feature importance DataFrame
            feature_importance_df = pd.DataFrame({
                'feature': transformed_feature_names,
                'coefficient': coefficients,
                'abs_coefficient': np.abs(coefficients)
            }).sort_values('abs_coefficient', ascending=False)
            
            print("Feature Importance Analysis (Top 15):")
            print("Rank | Feature Name           | Coefficient | Effect on LB Risk")
            print("-" * 65)
            
            top_features = feature_importance_df.head(15)
            for i, (idx, row) in enumerate(top_features.iterrows(), 1):
                direction = "Increases" if row['coefficient'] > 0 else "Decreases"
                print(f"{i:2d}   | {row['feature']:<20} | {row['coefficient']:>10.4f} | {direction}")
            
            # Analyze mechanism-based features specifically
            print(f"\nMechanism-Based CpG Features Analysis:")
            print("Gene Symbol | Original β | Model Coeff | Consistency | Original R²")
            print("-" * 70)
            
            mechanism_analysis = []
            for gene in self.available_mechanism_genes:
                orig_stats = self.cpg_mrna_pairs[gene]
                
                # Try to find this gene in the transformed features
                gene_coeff = None
                for feature_name, coeff in zip(transformed_feature_names, coefficients):
                    if gene in feature_name:
                        gene_coeff = coeff
                        break
                
                if gene_coeff is not None:
                    # Check consistency: both original and model should have same sign (negative)
                    consistency = "Consistent" if (orig_stats['beta'] < 0 and gene_coeff < 0) else "Inconsistent"
                    mechanism_analysis.append({
                        'gene': gene,
                        'original_beta': orig_stats['beta'],
                        'model_coeff': gene_coeff,
                        'original_r2': orig_stats['r_squared'],
                        'consistency': consistency
                    })
                    
                    print(f"{gene:<11} | {orig_stats['beta']:>9.3f} | {gene_coeff:>10.4f} | {consistency:<11} | {orig_stats['r_squared']:>9.3f}")
                else:
                    print(f"{gene:<11} | {orig_stats['beta']:>9.3f} | {'N/A':>10} | {'N/A':<11} | {orig_stats['r_squared']:>9.3f}")
            
            # Summary of mechanism-based features
            consistent_features = sum(1 for analysis in mechanism_analysis if analysis['consistency'] == 'Consistent')
            print(f"\nMechanism-Based Features Summary:")
            print(f"  Available features: {len(self.available_mechanism_genes)}")
            print(f"  Analyzed in model: {len(mechanism_analysis)}")
            print(f"  Consistent direction: {consistent_features}/{len(mechanism_analysis)}")
            
            if consistent_features / len(mechanism_analysis) >= 0.7:
                print("  Assessment: HIGH consistency with original CpG-mRNA regulatory direction")
            elif consistent_features / len(mechanism_analysis) >= 0.5:
                print("  Assessment: MODERATE consistency with original regulatory direction")
            else:
                print("  Assessment: LOW consistency - model may not reflect expected biology")
            
            # Save coefficients
            coef_file = os.path.join(self.output_dir, f"{self.prefix}feature_coefficients_{self.timestamp}.csv")
            feature_importance_df.to_csv(coef_file, index=False)
            print(f"\nFeature coefficients saved to: {coef_file}")
            
            return feature_importance_df, mechanism_analysis
            
        except Exception as e:
            print(f"Warning: Could not analyze feature importance: {e}")
            return pd.DataFrame(), []
    
    def save_results(self, multiple_splits_results, evaluation_results, mechanism_analysis):
        print("\n" + "=" * 50)
        print("STEP 10: SAVE RESULTS")
        print("=" * 50)
        
        model_file = os.path.join(self.output_dir, f"{self.prefix}trained_model_{self.timestamp}.joblib")
        joblib.dump(self.model, model_file)
        print(f"Model saved to: {model_file}")
        
        encoder_file = os.path.join(self.output_dir, f"{self.prefix}label_encoder_{self.timestamp}.joblib")
        joblib.dump(self.label_encoder, encoder_file)
        
        evaluation_file = os.path.join(self.output_dir, f"{self.prefix}evaluation_results_{self.timestamp}.json")
        combined_results = {
            'model_type': 'Mechanism-Based Feature Selection',
            'cpg_mrna_pairs_used': self.cpg_mrna_pairs,
            'available_mechanism_genes': self.available_mechanism_genes,
            'missing_mechanism_genes': self.missing_mechanism_genes,
            'evaluation_metrics': evaluation_results,
            'multiple_splits_analysis': multiple_splits_results,
            'mechanism_analysis': mechanism_analysis,
            'selected_features': self.feature_names,
            'timestamp': self.timestamp
        }
        
        with open(evaluation_file, 'w') as f:
            json.dump(combined_results, f, indent=4)
        print(f"Complete results saved to: {evaluation_file}")
        
        return True
    
    def run_complete_pipeline(self, clinical_path, dmr_path, methylation_path):
        print("\n" + "=" * 80)
        print("STARTING B12 PREDICTION - MECHANISM-BASED APPROACH")
        print("=" * 80)
        
        pipeline_start_time = datetime.now()
        
        try:
            if not self.load_data(clinical_path, dmr_path, methylation_path):
                print("ERROR: Data loading failed")
                return False
            
            if not self.prepare_combined_dataset():
                print("ERROR: Dataset preparation failed")
                return False
            
            multiple_splits_results = self.run_multiple_train_test_splits(n_iterations=20)
            
            self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
                self.X, self.y, test_size=0.2, random_state=42, stratify=self.y
            )
            
            if not self.create_preprocessing_pipeline():
                print("ERROR: Preprocessing pipeline creation failed")
                return False
            
            if not self.train_final_model():
                print("ERROR: Model training failed")
                return False
            
            evaluation_results = self.evaluate_final_model()
            
            feature_importance_df, mechanism_analysis = self.analyze_mechanism_feature_importance()
            
            if not self.save_results(multiple_splits_results, evaluation_results, mechanism_analysis):
                print("ERROR: Saving results failed")
                return False
            
            pipeline_end_time = datetime.now()
            pipeline_duration = pipeline_end_time - pipeline_start_time
            
            print("\n" + "=" * 80)
            print("PIPELINE COMPLETION SUMMARY - MECHANISM-BASED MODEL")
            print("=" * 80)
            print(f"Total duration: {pipeline_duration}")
            
            print(f"\nMechanism-Based Features Summary:")
            print(f"  Target CpG-mRNA pairs: {len(self.cpg_mrna_pairs)}")
            print(f"  Available in data: {len(self.available_mechanism_genes)}")
            print(f"  Missing from data: {len(self.missing_mechanism_genes)}")
            
            if self.missing_mechanism_genes:
                print(f"  Missing genes: {', '.join(self.missing_mechanism_genes)}")
            
            print(f"\nMultiple Splits Analysis Results:")
            splits_results = multiple_splits_results
            print(f"  Iterations completed: {splits_results['successful_iterations']}/{splits_results['n_iterations']}")
            
            if splits_results['performance_stats']['mean_auc']:
                mean_auc = splits_results['performance_stats']['mean_auc']
                std_auc = splits_results['performance_stats']['std_auc']
                print(f"  Mean AUC across splits: {mean_auc:.3f} ± {std_auc:.3f}")
                print(f"  Performance consistency: {splits_results['performance_stats']['consistency_level']}")
            
            print(f"\nFinal Model Performance:")
            auc = evaluation_results['test_metrics']['auc_roc']
            accuracy = evaluation_results['test_metrics']['accuracy']
            print(f"  Hold-out Test AUC: {auc:.4f}")
            print(f"  Hold-out Test Accuracy: {accuracy:.4f}")
            
            cv_auc = evaluation_results['cross_validation']['mean_cv_auc']
            cv_std = evaluation_results['cross_validation']['std_cv_auc']
            print(f"  Cross-validation AUC: {cv_auc:.4f} ± {cv_std:.4f}")
            
            # Mechanism consistency analysis
            if mechanism_analysis:
                consistent_count = sum(1 for analysis in mechanism_analysis if analysis['consistency'] == 'Consistent')
                total_analyzed = len(mechanism_analysis)
                consistency_rate = consistent_count / total_analyzed if total_analyzed > 0 else 0
                
                print(f"\nMechanism Consistency Analysis:")
                print(f"  Features with consistent regulatory direction: {consistent_count}/{total_analyzed} ({consistency_rate:.1%})")
                
                if consistency_rate >= 0.7:
                    print("  Assessment: HIGH biological consistency")
                elif consistency_rate >= 0.5:
                    print("  Assessment: MODERATE biological consistency")
                else:
                    print("  Assessment: LOW biological consistency")
            
            print(f"\nModel Type: Mechanism-based using predefined CpG-mRNA regulatory pairs")
            print(f"Advantage: Biologically informed features with known regulatory relationships")
            print(f"Key Insight: Performance reflects validity of mechanism-based hypothesis")
            print("=" * 80)
            
            return True
            
        except Exception as e:
            print(f"ERROR: Pipeline failed with exception: {str(e)}")
            self.logger.error(f"Pipeline failed with exception: {str(e)}")
            import traceback
            traceback.print_exc()
            return False


def main():
    """Main function to run the mechanism-based B12 prediction model"""
    
    clinical_path = "/Users/heweilin/Desktop/P056/7Clinical_data50.csv"
    dmr_path = "/Users/heweilin/Desktop/P056/4DNA_DMRs.csv"
    methylation_path = "/Users/heweilin/Desktop/P056_Code_2/Processed_Data/1_PD_PromoterRegion_CpGs.csv"
    output_dir = "/Users/heweilin/Desktop/P056_Code_3/Data"
    
    print("B12 Deficiency Prediction Model - Mechanism-Based Approach")
    print("=" * 70)
    print("Key Features:")
    print("- Uses 10 predefined CpG-mRNA negative regulatory pairs")
    print("- Biologically informed feature selection")
    print("- Multiple train-test splits for validation")
    print("- Batch correction for methylation data")
    print("- Clinical covariates + mechanism-based methylation features")
    print("- LASSO logistic regression with L1 regularization")
    print("=" * 70)
    print(f"Clinical data: {clinical_path}")
    print(f"Methylation data: {methylation_path}")
    print(f"Output directory: {output_dir}")
    print("=" * 70)
    
    model = B12PredictionMechanismBased(output_dir=output_dir, prefix="Mechanism_")
    
    success = model.run_complete_pipeline(clinical_path, dmr_path, methylation_path)
    
    if success:
        print("\n" + "=" * 80)
        print("MECHANISM-BASED MODEL PIPELINE COMPLETED SUCCESSFULLY!")
        print("=" * 80)
        print("Key Achievements:")
        print("- Successfully implemented mechanism-based feature selection")
        print("- Used predefined CpG-mRNA regulatory pairs as features")
        print("- Applied batch correction to methylation data")
        print("- Conducted multiple train-test splits for robust validation")
        print("- Analyzed consistency between original and model coefficients")
        print("- Final model trained with biologically meaningful features")
        print("\nThis approach tests whether known regulatory relationships")
        print("can predict B12 deficiency status effectively.")
        print("=" * 80)
    else:
        print("\nMECHANISM-BASED MODEL PIPELINE FAILED!")
        print("Check the log files for detailed error information.")
    
    return success


if __name__ == "__main__":
    np.random.seed(42)
    main()
```
